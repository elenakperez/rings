				δ_c  = δ_w  −0.24[‰/(°C)]  T
				
The quantity inside the square brackets contains the units of the $0.24$ constant. We can use this equation as an anomaly equation, where $δ_c$  , $δ_w$  and $T$ all represent changes from $20,000$ years ago to present.

### a) 
#### Solve an underdetermined least-squares problem for $δ_w$  and $T$. Use independent information as a first guess for the solution: $δ_w=−1‰$ from dilution by sea level rise, and $T=4.0°C$ from climate models. Pick a tapering parameter, $γ$, and rationalize your choice. State your solution for the two quantities and the corresponding standard errors.

```python
y = np.array([-1.9])
E = np.array([[1, -0.24]])
gamma_delta = 2 # initial guess for taper parameter
gamma_T = 10 # initial guess for taper parameter
x_0 = np.array([[-1], [-4]]) # prior info 

S = np.array([[1/gamma_delta, 0],[0, 1/gamma_T]])

x_tilde = np.linalg.inv((E.T@E + np.linalg.inv(S)))@(np.reshape(E.T@y, (2,1)) + np.linalg.inv(S)@x_0)

# store values for delta_w and T
delta_w = x_tilde[0]
T = x_tilde[1]

# compute delta_c from least squares estimation
delta_c = delta_w - (0.24*T)

print("The initial solution for δ_w = ",delta_w)
print("The initial solution for T = ",T)
print("The initial solution for δ_w =  δ_w - 0.24*T = ", delta_c)
```

### b)
##### Solve for the estimator, $E^†$, and show that it is a biased estimator.

To check if an estimator is unbiased $b=<\tilde x -x_{true}> = (E^†_{TW}E_{TW} − I)x_{true} + c  = 0$.

We know that $E^†=(E^TE + S^{-1})E^T$ where $S^{-1}\neq 0$ and that $c = (E^TE + I)^{-1}x_0$ where $x_0 \neq 0$.

Thus, we find that $c \neq 0$. As long the first time in the bias equation $b$ doesn't equal the negative of the second term then $b\neq=0$ which and therefore $E^†$ is a biased estimator.

## 1. Lagrange multipliers
Recall problem 4 on PSET 1. 

Oxygen-isotope ratios in calcite, $δ_c$, in a sediment core off Bermuda Rise show a change of $–1.9±0.2‰$ from $20,000$ years ago to the modern-day (Adkins et al. 2002). These ratios are determined by the oxygen-isotope ratio in seawater and the temperature in which the calcite formed. The equation that relates these quantities is, 

$δ_c=δ_w–0.24[‰/(°C)]T$. 

The quantity inside the square brackets is the units of the $0.24$ constant. We can use this equation as an anomaly equation, where $δ_c$, $δ_w$, and $T$ all represent changes from $20,000$ years ago to present. Is it possible to fit the Bermuda Rise observation of $δ_c= –1.9‰$ perfectly?

### a) 
#### Use Lagrange multipliers to solve for $δ_w$  and $T$ that fit the data point perfectly while minimizing the sum of squared deviations from the first guess of $δ_w=–1‰$ and $T=3°C$.

```python
'CREATE VARIABLES, VECTORS, AND MATRICES'
y = np.array([-1.9])
E = np.array([[1, -0.24]])
x_0 = np.array([[-1], [-3]]) # prior info 
gamma = 0.005 # taper parameter
S = np.array([[1/gamma, 0],[0, 1/gamma]]) 

'CONSTRUCT AN AUGMENTED COST FUNCTION FOR LAGRANGE MULTIPLIER METHOD'
print("J' = ((x_0-x).T)S(x_0-x) - 2(mu.T)(y-(Ex))")
print("")

print("μ = (ESE.T)^-1(y-Ex_0)")
mu = (1/(E@S@E.T))*(y-(E@x_0)) # lagrange multiplier
print("μ = ", mu)
print("")

print("x̃ = x_0 + (S*E.T*mu)")

x_tilde = x_0 + (S@E.T@mu)
print("x̃ = ", x_tilde)
print("")

# store values for delta_w and T
delta_w = x_tilde[0]
T = x_tilde[1]

'COMPUTE DELTA_C'
delta_c = delta_w - (0.24*T)

print("δ_w = ",delta_w)
print("T = ",T)
print("δ_w =  δ_w - 0.24*T = ", delta_c)
```

Using the Lagrange Multiplier method I found $\delta_c = -1.9$. This method predicted the solution with greater accuracy than the tapered, weighted least squares estimation I used in the problem #4 of pset #1.

### b) 
#### What is the value of the Lagrange multiplier? How do you interpret this value?

The value of the Lagrange multiplier $\mu = -0.00765885$. From the notes I know that $μ = ∂J′/∂y$, which is how the augmented cost function is changing with respect to y, where y are the observations and J is the cost function meant to minimize the distance between y and Ex, a.k.a. the noise in our problem. I interpret the value of $μ$ to mean the point in the solution space where $J'$ is minimized. 

## 2. Tapered, weighted least squares
Satellites typically go over a particular location on the Earth’s surface on a repeatable schedule. At crossover points where two satellite tracks cross, more observations are available but they are not evenly spaced in time. Consider (idealized) observations of sea surface salinity from two tracks of the Aquarius satellite, where "d" stands for days and salinity is in units of Absolute Salinity. The 1σ standard error of each observation is $0.1 g/kg$.

### a) 
#### Solve for the four coefficients of a third-order polynomial, $y=c_0+c_1 t+c_2 t^2+c_3 t^3$, that minimize the sum of squared misfits. Include their standard errors. Plot the data and the polynomial solution. Consider removing the mean of the observations and adding it back to the solution after performing calculations.

```python
'CREATE VARIABLES, VECTORS, AND MATRICES'

# sea surface salinity obs [absolute salinity]
y = np.array([[35.42],[35.31],[35.34],[35.12]])

# time [days]
t_1 = 0
t_2 = 0.1
t_3 = 7
t_4 = 7.1

# linear estimator 
E = np.array([[1,t_1,t_1**2,t_1**3],[1,t_2,t_2**2,t_2**3],[1,t_3,t_3**2,t_3**3],[1,t_4,t_4**2,t_4**3]])

'TO SOLVE FOR COEFFICIENTS OF THE 3-ORDER POLYNOMIAL SOLVE FOR x_tilde = E_dag*y'

# least squares estimator = E_dag, it is the matrix that transforms y to get x_tilde 
E_dag = (np.linalg.inv((E.T@E)))@E.T # E_dag = (E^T*E)^-1*E^T formula

# approximation of x_true values is x_tilde, i.e. for this problem the coefficients c_0, c_1, c_2, c_3 = x_tilde
x_tilde = E_dag@y

print('C = ', x_tilde.T[0])
```

The coefficients for the 3rd-order polynomial are $ C = [35.420, -1.162, 0.630, -0.066]$

```python
'TO FIND THE STANDARD ERROR, DERIVE THE SOLUTION COVARIANCE MATRIX C_xx AND TAKE THE SQUARE ROOT OF THE' 
'DIAGONALS OF THE MATRIX TO GET THE STANDARD ERROR. NOTE: C_nn IS THE NOISE COVARIANCE MATRIX'

# 1*sigma standard error of each obs = 0.1 [g/kg]
sigma = 0.1

C_nn = np.array([[sigma**2, 0, 0, 0],[0, sigma**2, 0, 0],[0, 0, sigma**2, 0],[0, 0, 0, sigma**2]])

C_xx = E_dag@C_nn@E_dag.T

sigma_x = np.sqrt(np.diag(C_xx)) # square root of the diagonals of solution covariance matrix is the standard error of the solution covariance

print('\u03C3_x = ',sigma_x)
```

The standard errors for the solution are $\sigma_x = [0.100, 1.455, 0.458, 0.041] $ 

```python
'PLOT THE DATA AND THE POLYNOMIAL SOLUTION'

t = [t_1, t_2, t_3, t_4] # time data points

# representative function of y
# y_poly = coeffs[0]*(t**0) + coeffs[1]*(t**1) + coeffs[2]*(t**2) + coeffs[3]*(t**3) 

coeffs = x_tilde
# x values for poly fit line
x = np.linspace(-1,8,100)

# generate values for poly fit line
fx = []
for i in range(len(x)):
    fx.append(coeffs[0]*(x[i]**0) + coeffs[1]*(x[i]**1) + coeffs[2]*(x[i]**2) + coeffs[3]*(x[i]**3))


# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('3rd-order polynomial solution for SSS observations', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(x,fx, alpha=0.8) # polynomial fit
ax.errorbar(t,y,color='r', yerr=0.1, fmt="o",markersize=6) # plot observational data with standard error bars

# axes formatting
ax.set_xlabel('Time [day]', fontweight='bold')
ax.set_ylabel('Sea Surface Salinity [g/kg]',fontweight='bold')
ax.set_ylim([34.75,37.15])
ax.set_xlim([-1,8])
ax.vlines([0,7],0,100,color='k',linestyle='--', alpha=0.1);
```

*Figure 1:* A $3^{rd}$-order polynomial solution for 4 observations of sea surface salinity from satellites. The red dots are the data observations with error bars. The blue line is the polynomial fit of the data. The dashed gray lines show 1 week between measurements.

## 3. Singular value decomposition and the projection theorem
An inverted echo sounder is placed on the seafloor and imperfectly measures sound travel times bouncing off the sea surface. The goal is to use the sound travel time to infer the compressibility $(x_1, x_2)$ of a two-layer ocean above. The units of compressibility are $dbar^{−1}$. At location 1, the echo sounder samples both layers equally. At location 2, the bottom layer is preferentially sampled due to a depression in the layer interface. At location 3, the top layer is preferentially sampled. 
The three observations yield the following equations in two unknowns $(x_1, x_2)$ , where units are given in the square brackets.
* $(1/2)[dbar s] x_1+(1/2)[dbar s] x_2+n_1=y_1$
* $(1/4)[dbar s] x_1+(3/4)[dbar s] x_2+n_2=y_2$
* $(3/4)[dbar s] x_1+(1/4)[dbar s] x_2+n_3=y_3$

The observations are, $y_1=5s±1s$, $y_2=4s±1s$, $y_3=7s±2s$ where the error bars are at the one standard deviation level.

### a) 

#### What is the particular SVD solution and the misfit to the data?

```python
import scipy
'CREATE VARIABLES, VECTORS, AND MATRICES'
y = np.array([[5],[4],[7]]) # observations
sigma_y = np.array([[1],[1],[2]]) # error bars, 1*sigma
E = np.array([[(1/2),(1/2)],[(1/4),(3/4)],[(3/4),(1/4)]])

# U = 
# V = 

# Lambda = U.T@E@V 

# x_tilde_p = V_k@(np.linalg.inv(Lambda_k))@(U_k.T)@y

# np.linalg.svd(E)[0]

(U,lambdas,Vh)  = np.linalg.svd(E) 

Lambda = np.diag(lambdas)

U = U[0:3,0:2]

(U @ Lambda @ Vh) # sanity check that this returns E, it does

x_tilde_p = (Vh.T)@(np.linalg.inv(Lambda))@(U.T@y)
print("x̃_p = ",x_tilde_p.T)
```

The particular SVD solution I found is $\tilde x_p = [8.333, 2.333]$. Since I have $E, \tilde x, y$ I can calculate $\tilde n$ since $\tilde n = E\tilde x - y$.

```python
n_tilde = E@x_tilde_p - y
print("ñ = ",n_tilde.T)
```

The misfits of this particular SVD solution is $\tilde n = [ 0.333, -0.167, -0.167]$.

### b) 
#### Confirm that the misfit of the particular SVD solution is normal to the range of E

```python
print("E^T*n = ", (E.T@n_tilde).T)
```

To show that the misfits are normal to the range of E we need to find $E^Tn = 0$. With Python I found that $E^Tn =  [-3.33e-15, -1.998e-15]$. This is essentially zero in Python, but is reported as an extremely small number due to the limited nature of significant digits that a computer can computationally store. Thus, $E^Tn = 0$ and the misfit of the particualar SVD solution is normal to the range of $E$.

```python
'PLOT THE DATA AND THE SOLUTION'

x = np.arange(0, 51, 1) # x values for plot

coeffs = x_tilde_p

# generate values for poly fit line
fx = []
for i in range(len(x)):
    fx.append((coeffs[0]*np.sin(2*np.pi*x[i]*1 / L) + coeffs[1]*np.sin(2*np.pi*x[i]*2 / L) + coeffs[2]*np.sin(2*np.pi*x[i]*3 / L) + coeffs[3]*np.sin(2*np.pi*x[i]*4 / L) + coeffs[4]*np.sin(2*np.pi*x[i]*5 / L) + coeffs[5]*np.sin(2*np.pi*x[i]*6 / L) + 
               coeffs[6]*np.cos(2*np.pi*x[i]*1 / L) + coeffs[7]*np.cos(2*np.pi*x[i]*2 / L) + coeffs[8]*np.cos(2*np.pi*x[i]*3 / L) +  coeffs[9]*np.cos(2*np.pi*x[i]*4 / L) + coeffs[10]*np.cos(2*np.pi*x[i]*5 / L) + coeffs[11]*np.cos(2*np.pi*x[i]*6 / L) + 
               coeffs[12]))

# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('New SVD Solution for SST Anomalies on R/V Armstrong', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(x,fx, alpha=0.8) # polynomial fit
ax.errorbar(d,y,color='r', yerr=0.1, fmt="o",markersize=6) # plot observational data with standard error bars

# axes formatting
ax.set_xlabel('Distance [km]', fontweight='bold')
ax.set_ylabel('Sea Surface Temperature Anomaly [deg C]',fontweight='bold')
# ax.set_ylim([-1,1])
ax.set_xlim([-1,51])
ax.hlines(y=0,xmin=-1,xmax=51,color='k',linestyle='--', alpha=0.1);
```
