```python
m_s = 2*10e30
m_e = 6*10e24
m_m = 7.3*10e22
r = 6.371*10e6
R_m = 3.825*10e8
D_m = 3.88871*10e8
R_s = 1.5*10e11
D_s = 1.5006371*10e11
G = 6.67*10e-11 

theta_array = (np.arange(0,361,3) * math.pi)/180

# f_t = sqrt(f_tx^2 + f_ty^2)
```

```python
# case 1
f_tx = ((2*G*m_e*m_m*r)/(D_m)**3)*np.cos(theta_array) + ((2*G*m_e*m_s*r)/(D_s**3))*np.cos(theta_array)
f_ty = -(((G*m_e*m_m*r)/(D_m)**3)*np.sin(theta_array)) - (((G*m_e*m_s*r)/(D_s**3))*np.sin(theta_array))
```

```python
# pset #2

u_g = np.arange(-1,4.1,0.01)
v_g = np.arange(-1,4.1,0.01)

z = np.arange(-1,4.1,0.01)
d = np.arange(-1,4.1,0.01)

# solution for u velocity
u_e = (u_g-np.exp(-z/d))*(u_g*np.cos(z/d)+v_g*np.sin(z/d))

# solution for v velocity
v_e = (v_g+np.exp(-z/d))*(u_g*np.sin(z/d)-v_g*np.cos(z/d))
```

```python
# surface ekman layer

fig,ax = plt.subplots()

ax.plot(u_e,z)
ax.plot(v_e,z)
ax.set_xlim([-1,1])
ax.set_ylim([0,4])
plt.axvline(x = 0, color='k')
```

```python
# bottom ekman layer

u_g = np.arange(-1,6.1,0.01)
v_g = np.arange(-1,6.1,0.01)
z = np.arange(-1, 6.1, 0.01) # (z+h)/delta

u_g = 1


c= u_g + u_g*-np.exp(-(1+np.sqrt(-1+0j))*z)

u_e = np.real(c)
v_e = np.imag(c)
```

```python

fig,ax = plt.subplots()

fig.set_dpi(100)

ax.plot(u_e,z,color='b', label='Cross pg (to x)')
ax.plot(v_e,z,color='r', label='Down pg (to y)')
ax.set_xlim([-0.7,1.2])
ax.set_ylim([0,6])
plt.axvline(x = 0, color='k',linestyle='--')
plt.axvline(x = 1, color='k',linestyle='--');

plt.suptitle('Bottom Ekman Layer', y=0.96, fontsize=17);
ax.set_xlabel('Velocity ['+r'$\frac{u}{U}, \frac{v}{U}$'+']', fontsize=15);
ax.set_ylabel('Non-dimensional Depth ['+r'$\frac{z+h}{\delta}$'+']', fontsize=15);
ax.legend(loc='upper left');
# ax.invert_xaxis()
```
