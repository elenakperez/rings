# Flat Bottom case
### Velocity 
$v(x,z) = v_{max}A(x)\cdot B(z)$

such that

$A(x=x_{core})=1$

$B(z=0)=1$

For now, we have a gaussian times a quadratic to represent an idealized jet velocity structure.

$v(x,z) = \frac{v_{max}}{H^2}e^{-\frac{1}{2}(\frac{x-b}{c})^2}(z+H)^2$

```python
import math 

def v_func(x, z, b, c, H, v_max):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Input:
        x (array)  : offshore position
        z (array)  : depth
        a (int)    : height of the curve's peak, i.e. magnitude of the jet velocity
        b (int)    : the position of the center of the peak, i.e. the location of the core of the jet
        c (int)    : the standard deviation, i.e. the width of the jet
        H (int)    : length scale at which jet decays
        
    Output:
        v (array)  : velocity of the shelf break jet as a function of x and z
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Ax = (np.exp(-0.5*((x-b)/c)**2))
    Bz = (z+H)**2
    return (v_max/(H**2)) * Ax * Bz

# x = np.linspace(0, 50, 50)
x = np.arange(0,300.1*10**3,0.1*10**3)
z = np.arange(-40.1, 0, 0.1)

X, Z = np.meshgrid(x, z)
# V = v_func(X, Z, b=150*10**3, c=10*10**3, H=100, v_max=-0.2) 
V = v_func(X, Z, b=150*10**3, c=10*10**3, H=40, v_max=-0.2) 

'PLOT'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
# fig.suptitle('Velocity Structure of Shelfbreak Jet (idealized)', fontsize=14, y=0.925, fontweight='bold')

im = ax.contourf(X, Z, V, levels=10);

# axes formatting
ax.set_xlabel('x, offshore distance [km]', fontweight='bold')
ax.set_ylabel('z, depth [m]',fontweight='bold');
# xtick formatter
locs, labels = plt.xticks()
plt.xticks(locs,locs/1000)
# ax.set_xticks([]);
# ax.invert_yaxis()
# ax.legend();
# ax.set_xlim([0,100])

# colorbar
cbaxes = inset_axes(ax, width="5%", height="100%", loc='lower left', bbox_to_anchor=(1.05, 0., 1, 1), bbox_transform=ax.transAxes, borderpad=0)
cb = plt.colorbar(im, cax = cbaxes)
cb.set_label('Velocity [m/s]',weight='bold')
```

### Density 
Derive density field from velocity field such that thermal wind is satisfied

$-f\frac{\partial v}{\partial z} = g\frac{\partial \rho}{\partial x}$

thus

$\frac{\partial v}{\partial z} = -\frac{g}{f}\frac{\partial \rho}{\partial x}$

Integrate with respect to $x$ to get the density field

$\rho'(x,z) = \int_{x'}^\infty (\frac{\partial v}{\partial z} = -\frac{g}{f}\frac{\partial \rho}{\partial x}) dx'$

...

$\rho'(x,z) = \frac{2f\rho_{ref}cv_{max}}{gH^2}\sqrt{\frac{\pi}{2}}(z+H)\Bigr[1 -erf\bigr(\frac{x-b}{c\sqrt{2}})\Bigr]$

```python
import scipy

def rho_func(x, z, b, c, H, rho_0, g, f, v_max):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Input:
        x (array)     : offshore position
        z (array)     : depth
        a (int)       : height of the curve's peak, i.e. magnitude of the jet velocity
        b (int)       : the position of the center of the peak, i.e. the location of the core of the jet
        c (int)       : the standard deviation, i.e. the width of the jet
        H (int)       : length scale at which jet decays
        rho_0 (int)   : open ocean reference density (e.g. 1000 kg/m^3)
        g (int)       : gravitational constant 
        f (int)       : coriolis parameter
        v_max (int)   : maximum velocity of SBJ at the core of the jet
        
    Output:
        rho (array)   : density as a function of x and z
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    return ((((2*f*rho_0*c*v_max)/(g*H**2))*(math.sqrt(math.pi/2))*(z+H)*(1-scipy.special.erf(((x-b)/(math.sqrt(2)*c))))))

x = np.arange(0,300.1*10**3,0.1*10**3) # meters = (km x 10**3)
z = np.arange(-100.1, 0, 0.1)

X, Z = np.meshgrid(x, z)
rho = rho_func(X, Z, b=150*10**3, c=10*10**3, H=100, rho_0=1000, g=9.8, f=(10**-4), v_max = -0.2)

'PLOT'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
# fig.suptitle('rho_prime(x,z)', fontsize=14, y=0.925, fontweight='bold')

# plot
im = ax.contourf(X, Z, rho);

# axes formatting
ax.set_xlabel('x, offshore distance [km]', fontweight='bold')
ax.set_ylabel('z, depth [m]',fontweight='bold');
# xtick formatter
locs, labels = plt.xticks()
plt.xticks(locs,locs/1000)
# ax.invert_yaxis()
# ax.legend();
# ax.set_xlim([0,100])

# colorbar
cbaxes = inset_axes(ax, width="5%", height="100%", loc='lower left', bbox_to_anchor=(1.05, 0., 1, 1), bbox_transform=ax.transAxes, borderpad=0)
cb = plt.colorbar(im, cax = cbaxes)
cb.set_label('Density Anomaly [kg^3/m]',weight='bold')
```

### Sea Level 
Derive sea level field from velocity field such that geostrophy is satisfied

$-fv(x, z=0) = -g\zeta_x$

thus

$v = \frac{g}{f}\zeta_x$

Integrate with respect to $x$ to get the sea level field

$\int_{x'}^\infty v dx = \frac{g}{f}(\zeta(\infty)-\zeta(x))$

where $\zeta(\infty) = 0$ therefore

$\zeta(x) = -\frac{f}{g} \int_{x'}^\infty v dx$

where $v = v(x, z = 0)$

Solving for $\zeta(x)$ we find $\zeta(x) = -\frac{fH^2}{g}ac\sqrt{\frac{\pi}{2}}erf\bigr(\frac{b-x}{\sqrt{2}c}\bigr)$

where $a = \frac{v_{max}}{H^2}$ so

$\zeta(x) = -\frac{fv_{max}c}{g}\sqrt{\frac{\pi}{2}}[1-erf\bigr(\frac{x-b}{\sqrt{2}c}\bigr)]$


```python

def zeta_func(x, b, c, H, g, f, v_max):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Input:
        x (array)     : offshore position
        a (int)       : height of the curve's peak, i.e. magnitude of the jet velocity
        b (int)       : the position of the center of the peak, i.e. the location of the core of the jet
        c (int)       : the standard deviation, i.e. the width of the jet
        H (int)       : length scale at which jet decays
        g (int)       : gravitational constant 
        f (int)       : coriolis parameter
        v_max (int)   : maximum velocity of SBJ at the core of the jet
        
    Output:
        zeta (array)  : sea level as a function of x
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    return (((-(f*v_max*c)/g)*(math.sqrt(math.pi/2)))*(1-scipy.special.erf(((x-b)/(math.sqrt(2)*c)))))

x = np.arange(0,300.1*10**3,0.1*10**3) # meters = (km x 10**3)

zeta = zeta_func(x, b=150*10**3, c=10*10**3, H=100, g=9.8, f=(10**-4), v_max = -0.2)

'PLOT'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
# fig.suptitle('zeta(x,z)', fontsize=14, y=0.925, fontweight='bold')

# plot
im = ax.plot(x, zeta);

# axes formatting
ax.set_xlabel('x, offshore distance [km]', fontweight='bold')
ax.set_ylabel('zeta, sea leve height [m]',fontweight='bold');
# xtick formatter
locs, labels = plt.xticks();
plt.xticks(locs,locs/1000);
# ax.invert_yaxis()
# ax.legend();
ax.set_xlim([0,301*1000]);

```

### Transport 
Derive transport from velocity field 

$Q = \int \int v(x,z) dz dx$

Our bounds are from onshore $x=0$ to offshore $x=+\infty$ and the bottom $z=-H(x)$ to the top $z=0$. 
Note that the "top" is considered $z=0$ because we are using the linear free surface approximation which means 
that when we integrate over the full water column $z=-H(x)$ to $z=\zeta(x)$ that the variations in $\zeta(x)$
are so miniscule that integrating to $z=0$ is a really good approximation.

$Q = \int_{0}^{+\infty} \int_{-H_0}^{0} \frac{v_{max}}{H_0^2}e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2}\left(z+H_0\right)^2 dz dx$

Integrate with respect to $z$ first

$Q = \int \int v(x,z) dz dx = \frac{v_{max}}{H_0^2} \int_{0}^{+\infty} e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2}\frac{1}{3}H_0^3 dx$

Integrate with respect to $x$

$= \frac{1}{3}v_{max} H_0 c \sqrt{\frac{\pi}{2}} \left[1 - erf\left(\frac{-b}{c\sqrt{2}}\right) \right]$

Thus

$Q = \frac{v_{max}H_0c}{3} \sqrt{\frac{\pi}{2}} \left[1 - erf\left(\frac{-b}{c\sqrt{2}}\right) \right]$

for the flat bottom case

```python

def Q_func(b, c, H, v_max):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Input:
        x (array)     : offshore position
        b (int)       : the position of the center of the peak, i.e. the location of the core of the jet
        c (int)       : the standard deviation, i.e. the width of the jet
        H (int)       : depth of the flat bottom
        v_max (int)   : maximum velocity of SBJ at the core of the jet
        
    Output:
        Q (array)     : transport of the jet structure
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    return ((v_max*H*c)/3*(math.sqrt(math.pi/2)))*(1-scipy.special.erf(((-b)/(math.sqrt(2)*c))))

x = np.arange(0,300.1*10**3,0.1*10**3) # meters = (km x 10**3)

Q = Q_func(b=150*10**3, c=10*10**3, H=100, v_max = -0.2)

print('Transport Q for flat bottom case where b=150km, c=10km, H_0=100m, and v_max=-0.2m/s is Q = ', Q,'m^3/s')
print('That is ', Q/10**6, ' Sv of transport')
```

# Simple Bathymetry case
### Bathymetry 
\begin{equation}
H(x)=
    \begin{cases}
        H_0 e^{\frac{x}{\beta}} & \text{for } 0 < x < x_b\\
        H_d & \text{for } x_b < x\\
    \end{cases}
\end{equation}

Where $\beta = \frac{x_b}{ln(\frac{H_d}{H_0})}$

Some reasonable values for $H_0$ and $H_d$ might be $H_0 = 40 m$ and $H_d = 4000 m$. A reasonable for $x_b$, assuming $x=0$ is the coastline and distance is measured by $km$ in the offshore distance, could be $x_b = 70 km$. 

```python
# H_0 = -40 # m
# H_d = -4000 # m
# x_b = 70*10**3 # km
# beta = x_b/(np.log(H_d/H_0))

x = np.arange(0,300.1*10**3,0.1*10**3) # meters = (km x 10**3)
z = np.arange(-4000.1, 0, 1)

H_0 = 40 # m
H_d = 4000 # m
x1 = 150*10**3; H1 = 80 # parameters from Forsyth et al., 2021 fig 6 (right panel)
x2 = 200*10**3; H2 = 1000 # m
beta = (x2-x1)/(np.log(H2/H1))
alpha = np.exp(np.log(H2)-x2/beta)
x_b = x[(np.where(alpha*np.exp(x/beta)>H_d))[0][0]]

# piecewise function H(x)
Hx = np.piecewise(x, [(0 < x) & (x < x_b), (x >= x_b)], [lambda x : alpha*np.exp(x/beta), H_d])

'PLOT'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(60)
fig.suptitle('Simple Bathymetry', fontsize=14, y=0.925, fontweight='bold')

# plot
plt.plot(x, -Hx, color='k');

# axes formatting
ax.set_xlabel('x (offshore)', fontweight='bold')
ax.set_ylabel('depth [m]',fontweight='bold');
# xtick formatter
locs, labels = plt.xticks();
plt.xticks(locs,locs/1000);
# ax.invert_yaxis()
# ax.legend();
# ax.set_xlim([0,100])
ax.set_xlim([0,301*1000]);
```

### Velocity 
$v(x,z) = v_{max}A(x)\cdot B(z)$

such that

$A(x=x_{core})=1$

$B(z=0)=1$

For now, we have a gaussian times a quadratic to represent an idealized jet velocity structure.

$v(x,z) = \frac{v_{max}}{H(x)^2}e^{-\frac{(x-b)^2}{2c^2}}(z+H(x))^2$

```python
x1 = 150*10**3; H1 = 80 # parameters from Forsyth et al., 2020
x2 = 200*10**3; H2 = 1000 # m
beta = (x2-x1)/(np.log(H2/H1))
alpha = np.exp(np.log(H2)-x2/beta)
H_d = 4000 # m
x_b = x[(np.where(alpha*np.exp(x/beta)>H_d))[0][0]]
```

```python
def v_hx_func(x, z, b, c, Hx, v_max):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Input:
        x (array)  : offshore position
        z (array)  : depth
        a (int)    : height of the curve's peak, i.e. magnitude of the jet velocity
        b (int)    : the position of the center of the peak, i.e. the location of the core of the jet
        c (int)    : the standard deviation, i.e. the width of the jet
        Hx (array) : depth of the bathymetry
        
    Output:
        v (array)  : velocity of the shelf break jet as a function of x and z
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Ax = (np.exp(-((x-b)**2)/(2*c**2)))
    Bz = (z+Hx)**2
    return ((v_max/(Hx**2)) * Ax * Bz)

# x = np.linspace(0, 50, 50)
x = np.arange(0,300.1*10**3,1*10**3)
z = np.arange(-4000.1, 0, 1)

# bathymetry
H_0 = 40 # m
H_d = 4000 # m
x1 = 150*10**3; H1 = 80 # parameters from Forsyth et al., 2020
x2 = 200*10**3; H2 = 1000 # m
beta = (x2-x1)/(np.log(H2/H1))
alpha = np.exp(np.log(H2)-x2/beta)
x_b = x[(np.where(alpha*np.exp(x/beta)>H_d))[0][0]]

# piecewise function H(x)
Hx = np.piecewise(x, [(0 < x) & (x < x_b), (x >= x_b)], [lambda x : alpha*np.exp(x/beta), H_d])

X, Z = np.meshgrid(x, z)
V = v_hx_func(X, Z, b=150*10**3, c=10*10**3, Hx=Hx, v_max=-0.2)

# make mask for bathymetry
mask = np.nan*np.zeros((len(x), len(z)))     
for i in range(len(x)):
    for j in range(len(z)):
        mask[i,j] = 1
        if abs(z[j])>abs(Hx[i]):
            mask[i,j] = np.nan

'PLOT'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
# fig.suptitle('Velocity Structure of Shelfbreak Jet (idealized)', fontsize=14, y=0.925, fontweight='bold')

# plot
# ax.plot(x,Ax,label='A(x)') 
# ax.plot(z,Bz,label='B(z)')
# ax.plot(x,x*z)  

im = ax.contourf(X, Z, V*mask.T, levels=10);
plt.plot(x, -Hx, color='k');

# axes formatting
ax.set_xlabel('x, offshore distance [km]', fontweight='bold')
ax.set_ylabel('z, depth [m]',fontweight='bold');
# xtick formatter
locs, labels = plt.xticks();
plt.xticks(locs,locs/1000);
# ax.invert_yaxis()
# ax.legend();
ax.set_xlim([0,200*10**3])
ax.set_ylim([-200,0]);

# colorbar
cbaxes = inset_axes(ax, width="5%", height="100%", loc='lower left', bbox_to_anchor=(1.05, 0., 1, 1), bbox_transform=ax.transAxes, borderpad=0)
cb = plt.colorbar(im, cax = cbaxes)
cb.set_label('Velocity [m/s]',weight='bold')
```

### Density 
$\rho'(x,z) = \frac{2f\rho_{0}v_{max}}{g} \left( \frac{z}{H_0^2}e^{\frac{2}{\beta}\left(\frac{c^2}{\beta}-b\right)} c\sqrt{\frac{\pi}{2}} \left[erf\left(\frac{x_b + \left(\frac{2c^2}{\beta} -b \right)}{c\sqrt{2}}\right) - erf\left(\frac{x + \left(\frac{2c^2}{\beta} -b \right)}{c\sqrt{2}}\right) \right] \right.\\
            +\frac{1}{H_0}e^{\frac{1}{\beta}\left(\frac{c^2}{\beta}-b\right)} c\sqrt{\frac{\pi}{2}} \left[erf\left(\frac{x_b + \left(\frac{c^2}{\beta} -b \right)}{c\sqrt{2}}\right) - erf\left(\frac{x + \left(\frac{c^2}{\beta} -b \right)}{c\sqrt{2}}\right)\right]\\
            +\left.\frac{(z+H_d)}{H_d^2} c\sqrt{\frac{\pi}{2}} \left[ 1 -erf\bigr(\frac{x_b-b}{c\sqrt{2}}) \right]\right)$

```python
def rho_hx_func(x, z, b, c, H_0, H_d, x1, x2, H1, H2, rho_0, g, f, v_max):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Input:
        x (array)     : offshore position
        z (array)     : depth
        a (int)       : height of the curve's peak, i.e. magnitude of the jet velocity
        b (int)       : the position of the center of the peak, i.e. the location of the core of the jet
        c (int)       : the standard deviation, i.e. the width of the jet
        H_0 (int)     : initial depth of the shelf
        H_d (int)     : depth of the open ocean bottom
        x_b (int)     : distance offshore where slope transitions to flat bottom
        rho_0 (int)   : open ocean reference density (e.g. 1000 kg/m^3)
        g (int)       : gravitational constant 
        f (int)       : coriolis parameter
        v_max (int)   : maximum velocity of SBJ at the core of the jet
        leq_xb (Bool) : if True, x is less than or equal to x_b
        
    Output:
        rho (array)   : density as a function of x and z
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    # call x, z np arrays new name so i don't have to change the x,z in my rho equation to X,Z 
    x_og = x
    z_og = z
    
    # bathymetry
    beta = (x2-x1)/(np.log(H2/H1))
    alpha = np.exp(np.log(H2)-x2/beta)
    x_b = x[(np.where(alpha*np.exp(x/beta)>H_d))[0][0]]

    # for 0 < x < x_b
    x_less_xb = x_og[x_og<=x_b]
    x, z = np.meshgrid(x_less_xb, z_og)
    
    # break into prefactors by term 
    pf1 = (1/alpha**2)*np.exp((2/beta)*((c**2/beta)-b))*(c*(math.sqrt(math.pi/2)))
    pf2 = ((1/alpha)*np.exp((1/beta)*((c**2/beta)-b))*(c*(math.sqrt(math.pi/2))))
    pf3 = ((1/H_d**2)*(c*(math.sqrt(math.pi/2))))
    
    # break into terms
    term1 = z*((scipy.special.erf(((x_b+(((2*(c**2))/beta)-b))/(math.sqrt(2)*c))))-(scipy.special.erf(((x+(((2*(c**2))/beta)-b))/(math.sqrt(2)*c)))))
    term2 = ((scipy.special.erf(((x_b+((((c**2))/beta)-b))/(math.sqrt(2)*c))))-(scipy.special.erf(((x+((((c**2))/beta)-b))/(math.sqrt(2)*c)))))
    term3 = (z+H_d)*(1-(scipy.special.erf(((x_b-b)/(math.sqrt(2)*c)))))

    # density over variable topography
    rho_hx = ((2*f*rho_0*v_max)/g)*(pf1*term1 + pf2*term2 + pf3*term3)

    # for x>x_b
    x_greater_xb = x_og[x_og>x_b]
    x, z = np.meshgrid(x_greater_xb, z_og)

    term3 = (z+H_d)*(1-(scipy.special.erf(((x_b-b)/(math.sqrt(2)*c)))))

    # density over flat bottom
    rho_h = ((2*f*rho_0*v_max)/g)*(pf3*term3)

    # concatenate x<x_b and x>x_b
    rho_hx = np.concatenate((rho_hx,rho_h),axis=1)

    return rho_hx

# define x, z
x = np.arange(0,300.1*10**3,1*10**3)
z = np.arange(-4000.1, 0, 1)

# create x,z grid 
X, Z = np.meshgrid(x, z)

# bathymetry
H_0 = 40 # m
H_d = 4000 # m
x1 = 150*10**3; H1 = 80 # parameters from Forsyth et al., 2020
x2 = 200*10**3; H2 = 1000 # m
beta = (x2-x1)/(np.log(H2/H1))
alpha = np.exp(np.log(H2)-x2/beta)
x_b = x[(np.where(alpha*np.exp(x/beta)>H_d))[0][0]]

# piecewise function H(x)
Hx = np.piecewise(x, [(0 < x) & (x < x_b), (x >= x_b)], [lambda x : alpha*np.exp(x/beta), H_d])

# make mask for bathymetry
mask = np.nan*np.zeros((len(x), len(z)))     
for i in range(len(x)):
    for j in range(len(z)):
        mask[i,j] = 1
        if abs(z[j])>abs(Hx[i]):
            mask[i,j] = np.nan

# calculate density field by calling rho function
rho_hx = rho_hx_func(x, z, b=150*10**3, c=10*10**3, H_0=40, H_d=4000, x1=150*10**3, x2=200*10**3, H1=80, H2=1000, rho_0=1000, g=9.8, f=(10**-4), v_max = -0.2)

'PLOT'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
fig.suptitle('rho_prime(x,z)', fontsize=14, y=0.925, fontweight='bold')

# plot
im = ax.contourf(X, Z, rho_hx*mask.T, levels=10);
plt.plot(x, -Hx, color='k');

# axes formatting
ax.set_xlabel('x, offshore distance [km]', fontweight='bold')
ax.set_ylabel('z, depth [m]',fontweight='bold');
# xtick formatter
locs, labels = plt.xticks();
plt.xticks(locs,locs/1000);
# ax.invert_yaxis()
# ax.legend();
# ax.set_xlim([0,200*10**3])
ax.set_ylim([-150,0]);

# colorbar
cbaxes = inset_axes(ax, width="5%", height="100%", loc='lower left', bbox_to_anchor=(1.05, 0., 1, 1), bbox_transform=ax.transAxes, borderpad=0)
cb = plt.colorbar(im, cax = cbaxes)
cb.set_label('Density Anomaly [kg^3/m]',weight='bold')
```

### Sea Level 
$\zeta(x) = -\frac{fv_{max}c}{g} \sqrt{\frac{\pi}{2}}\left[1 - erf\left(\frac{x-b}{c\sqrt{2}}\right)\right]$

```python

def zeta_func(x, b, c, H, g, f, v_max):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Input:
        x (array)     : offshore position
        a (int)       : height of the curve's peak, i.e. magnitude of the jet velocity
        b (int)       : the position of the center of the peak, i.e. the location of the core of the jet
        c (int)       : the standard deviation, i.e. the width of the jet
        H (int)       : length scale at which jet decays
        g (int)       : gravitational constant 
        f (int)       : coriolis parameter
        v_max (int)   : maximum velocity of SBJ at the core of the jet
        
    Output:
        zeta (array)  : sea level as a function of x
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    return (((-(f*v_max*c)/g)*(math.sqrt(math.pi/2)))*(1-scipy.special.erf(((x-b)/(math.sqrt(2)*c)))))

x = np.arange(0,300.1*10**3,0.1*10**3) # meters = (km x 10**3)

zeta = zeta_func(x, b=150*10**3, c=10*10**3, H=100, g=9.8, f=(10**-4), v_max = -0.2)

'PLOT'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
fig.suptitle('zeta(x,z)', fontsize=14, y=0.925, fontweight='bold')

# plot
im = ax.plot(x, zeta);

# axes formatting
ax.set_xlabel('x, offshore distance [km]', fontweight='bold')
ax.set_ylabel('zeta, sea leve height [m]',fontweight='bold');
# xtick formatter
locs, labels = plt.xticks();
plt.xticks(locs,locs/1000);
# ax.invert_yaxis()
# ax.legend();
ax.set_xlim([0,301*1000]);

```

### Transport 
Derive transport from velocity field 

$Q = \int \int v(x,z) dz dx$

Our bounds are from onshore $x=0$ to offshore $x=+\infty$ and the bottom $z=-H(x)$ to the top $z=0$. 
Note that the "top" is considered $z=0$ because we are using the linear free surface approximation which means 
that when we integrate over the full water column $z=-H(x)$ to $z=\zeta(x)$ that the variations in $\zeta(x)$
are so miniscule that integrating to $z=0$ is a really good approximation.

$Q = \int_{0}^{+\infty} \int_{-H(x)}^{0} \frac{v_{max}}{H(x)^2}e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2}\left(z+H(x)\right)^2 dz dx$

Integrate with respect to $z$ first

$Q = \frac{1}{3}  v_{max}  \int_{0}^{+\infty} \frac{1}{H(x)^2} e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2} H(x)^3  dx$

Recall that $H(x)$ is piecewise so we can break this into 2 integrals where $H(x) = \alpha e^{\frac{x}{\beta}}$ for $0 < x < x_b$ and $H(x) = H_d$ for $x > x_b$

$= \frac{1}{3} v_{max} \left[ \int_{0}^{x_b} \alpha e^{\frac{x}{\beta}} e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2} dx + \int_{x_b}^{+\infty} H_d e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2}  dx \right]$

Let's consider the first integral first

$\int_{0}^{x_b} \alpha e^{\frac{x}{\beta}} e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2} dx$

After completing the square and doing a bunch of algebra we get:

$\int_{0}^{x_b} \alpha e^{\frac{x}{\beta}} e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2} dx = \alpha e^{\frac{1}{2\beta}\left(2b + \frac{c^2}{\beta}\right)}c\sqrt{\frac{\pi}{2}} \left[erf\left(\frac{x_b-(b+\frac{c^2}{\beta})}{c\sqrt{2}}\right) - erf\left(\frac{-(b+\frac{c^2}{\beta})}{c\sqrt{2}}\right)\right]$

Now let's get the second integral term. This one is easier.

$\int_{x_b}^{+\infty} H_d e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2}  dx$

Ingretate w.r.t. $x$, we've done this integral many times so we know it well by now to be

$\int_{x_b}^{+\infty} H_d e^{-\frac{1}{2}\left(\frac{x-b}{c}\right)^2}  dx = H_d c \sqrt{\frac{\pi}{2}} \left[1 - erf\left(\frac{x_b - b}{c\sqrt{2}}\right)\right]$

Plug what we've found for the first and second integrals back into our piecewise, 2-part integral to get our expression for $Q$

$Q = \frac{1}{3} v_{max} c\sqrt{\frac{\pi}{2}} \left( \alpha e^{\frac{1}{2\beta}\left(2b + \frac{c^2}{\beta}\right)} \left[erf\left(\frac{x_b-(b+\frac{c^2}{\beta})}{c\sqrt{2}}\right) - erf\left(\frac{-(b+\frac{c^2}{\beta})}{c\sqrt{2}}\right)\right] + H_d \left[1 - erf\left(\frac{x_b - b}{c\sqrt{2}}\right)\right] \right)$

```python

def Q_hx_func(alpha, beta, x_b, b, c, H_d, v_max):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    Input:
        x_b (int)     : distance offshore where slope transitions to flat bottom
        b (int)       : the position of the center of the peak, i.e. the location of the core of the jet
        c (int)       : the standard deviation, i.e. the width of the jet
        H_0 (int)     : initial depth of the shelf
        H_d (int)     : depth of the flat bottom offshore
        v_max (int)   : maximum velocity of SBJ at the core of the jet
        
    Output:
        Q (array)     : transport of the jet structure
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    return (v_max*c)/3*(math.sqrt(math.pi/2))*((alpha*np.exp((1/(2*beta))*(2*b + (c**2/beta))))*((scipy.special.erf(((x_b-(b+(c**2/beta)))/(math.sqrt(2)*c))))-(scipy.special.erf(((-b+(c**2/beta))/(math.sqrt(2)*c))))) + H_d*((1-scipy.special.erf(((x_b-b)/(math.sqrt(2)*c))))))

# bathymetry
x1 = 150*10**3; H1 = 80 # parameters from Forsyth et al., 2020
x2 = 200*10**3; H2 = 1000 # m
beta = (x2-x1)/(np.log(H2/H1))
alpha = np.exp(np.log(H2)-x2/beta) 
H_d = 4000 # m
x_b = x[(np.where(alpha*np.exp(x/beta)>H_d))[0][0]]

Q_hx = Q_hx_func(alpha = alpha, beta = beta, x_b = x_b, b=150*10**3, c=10*10**3, H_d=4000, v_max = -0.2)

print('Transport Q for varaible topography case where x_b = 227.5km, b=150km, c=10km, H_0=100m, and v_max=-0.2m/s is Q = ', Q_hx,'m^3/s')
print('That is ', Q_hx/10**6, ' Sv of transport')
```
