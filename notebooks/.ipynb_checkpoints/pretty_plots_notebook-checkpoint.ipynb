```python
import seaborn as sns
sns.set()

zone_eddies_formations = pd.read_pickle('/Users/elenaperez/Desktop/chatts/data/pd_dataframes/zone_eddies_formations.pkl')


min_year = int(min(zone_eddies_formations['year']))
max_year = int(max(zone_eddies_formations['year']))

pre_2000 = zone_eddies_formations[zone_eddies_formations['year']<2000] # mean rings count pre-2000
post_2000 = zone_eddies_formations[zone_eddies_formations['year']>=2000] # mean ring count post-2000

fig,ax = plt.subplots()
fig.set_dpi(100)
fig.suptitle('Anti-cyclonic vs. Cyclonic Eddy Formations (1993-2020)', fontsize=11, y=0.935)

# anticyclones
ax.plot(zone_eddies_formations['year'].iloc[0:27],zone_eddies_formations['anticyclonic'].iloc[0:27],'-o',color='r')
ax.plot([min_year,2000],[pre_2000['anticyclonic'].mean(),pre_2000['anticyclonic'].mean()], color='r', linestyle='-'); # pre_2000 average
ax.plot([2000,max_year],[post_2000['anticyclonic'].mean(),post_2000['anticyclonic'].mean()], color='r', linestyle='-'); # post_2000 average

# cyclones
ax.plot(zone_eddies_formations['year'].iloc[0:27],zone_eddies_formations['cyclonic'].iloc[0:27],'-o',color='b')
ax.plot([min_year,2000],[pre_2000['cyclonic'].mean(),pre_2000['cyclonic'].mean()], color='b', linestyle='-'); # pre_2000 average
ax.plot([2000,max_year],[post_2000['cyclonic'].mean(),post_2000['cyclonic'].mean()], color='b', linestyle='-'); # post_2000 average

# axes formatting
ax.set_xlabel('Time')
ax.set_ylabel('# annual eddy formations');
```

```python
min_year_ccr = int(min(zone_ccr_formations['year']))
max_year_ccr = int(max(zone_ccr_formations['year']))

min_year_wcr = int(min(zone_wcr_formations['year']))
max_year_wcr = int(max(zone_wcr_formations['year']))

pre_2000_ccr = zone_ccr_formations[zone_ccr_formations['year']<2000] # mean rings count pre-2000
post_2000_ccr = zone_ccr_formations[zone_ccr_formations['year']>=2000] # mean ring count post-2000

pre_2000_wcr = zone_wcr_formations[zone_wcr_formations['year']<2000] # mean rings count pre-2000
post_2000_wcr = zone_wcr_formations[zone_wcr_formations['year']>=2000] # mean ring count post-2000

fig,ax = plt.subplots()
fig.set_dpi(175)
fig.suptitle('WCR vs. CCR Formations ('+str(int(min(zone_ccr_formations['year'])))+' - '+str(int(max(zone_ccr_formations['year'])))+')', y=0.935, fontsize=11);

# anticyclones
ax.plot(zone_wcr_formations['year'],zone_wcr_formations['all_zones'],'-o',color='r')
ax.plot([min_year_wcr,2000],[pre_2000_wcr['all_zones'].mean(),pre_2000_wcr['all_zones'].mean()], color='r', linestyle='-'); # pre_2000 average
ax.plot([2000,max_year_wcr],[post_2000_wcr['all_zones'].mean(),post_2000_wcr['all_zones'].mean()], color='r', linestyle='-'); # post_2000 average

# cyclones
ax.plot(zone_ccr_formations['year'],zone_ccr_formations['all_zones'],'-o',color='b')
ax.plot([min_year_ccr,2000],[pre_2000_ccr['all_zones'].mean(),pre_2000_ccr['all_zones'].mean()], color='b', linestyle='-'); # pre_2000 average
ax.plot([2000,max_year_ccr],[post_2000_ccr['all_zones'].mean(),post_2000_ccr['all_zones'].mean()], color='b', linestyle='-'); # post_2000 average

# axes formatting
ax.set_xlabel('Time')
ax.set_ylabel('# annual ring formations');

# custom legend
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='r', lw=4), Line2D([0], [0], color='b', lw=4)]
ax.legend(custom_lines, ['WCR', 'CCR']);
```
