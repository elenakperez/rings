```python
#imports
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
#get the data
data_q1 = scipy.io.loadmat('C://Users/2021/Downloads/CODE.mat')['arr']
data_q3 = scipy.io.loadmat('C://Users/2021/Downloads/NewportPortlandSealevel.mat')['arr']
```

```python
#separate the data
time = data_q1[:,[0]].T
u = data_q1[:,1:6].T
v = data_q1[:,6:11].T
u_wind = data_q1[:,11:12].T
v_wind = data_q1[:,12:13].T

time_q3 = data_q3[:,[0]].T
np_sl = data_q3[:,[1]].T
pl_sl = data_q3[:,[2]].T
```

```python
#1a
#covariance method
y = v #the y is set to the v vectors of the data
y = y - y.mean(axis = 1, keepdims = True) #subtract the mean
C_yy = (y @ y.T) / y.shape[1] #calculate the covariance matrix
eig_val, eig_vec = np.linalg.eig(C_yy) #get the eigenvalues and eigenvectors
U_cov_method = eig_vec #the U matrix represents the eigenvetors of C_yy

#SVD method
y = v #the y is set to the v vectors of the data
y = v - v.mean(axis = 1, keepdims = True) #subtract the mean
U, A, V = np.linalg.svd(y) #do the singular value decomposition (A = lambda)
U_svd_method = U

#you can see that the results are equilivent
print(f'U_cov_method: \n {U_cov_method}')
print(f'U_svd_method: \n {U_svd_method}')
```

```python
#1b
#plot the first EOF (as a function of station number (x=1=station1=5m, etc.))
EOF1 = U[:,0] #get the first EOF (from 1a)
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5], EOF1)
ax.set_xlabel('station #')
ax.set_ylabel('cm/s')
ax.set_xticks([1,2,3,4,5])
ax.set_title('EOF1')


#plot the timeseries at each station
fig, ax = plt.subplots(5)
for i in range(5):
    ax[i].plot(time[0], v[i])
    ax[i].set_xlabel('time (1992 year days)')
    ax[i].set_ylabel('v (cm/s)')
    ax[i].set_ylim(-50,50)
    ax[i].set_title(f'v at station {i}')
    
"""
We can see the spatial pattern of EOF1 when plotting. 
The magnitude of values of EOF1 decrease at each station.
This is reasonable and expected becuase the variability of the ocean and its currents decreases and becomes more unifrom with depth.
This can be observed by looking at the v timeseries at each station. 
Here, you can see that the magnitude and variability decreases with depth at each successive station.
"""

#calculate the percentage of variance explained by each EOF
U, A, V = np.linalg.svd(y) #the A variable (lambda) represents the eigenvalues
percents_cumulative = []
percents_individual = []
for i in range(A.size):
    percent_cumulative = (np.sum(A[i:]**2)) / (np.sum(A[:]**2))
    percent_individual = (A[i]**2) / (np.sum(A[:]**2))
    percents_cumulative.append(percent_cumulative)
    percents_individual.append(percent_individual)

#show the percent of variance that is explained by the first EOF
print(f'percent variance explained by EOF1: {round(percents_individual[0] * 100, 1)}%')


```

```python
#1c
U, A, V = np.linalg.svd(y, full_matrices = False) #do the svd

#V is a 5x485 matrix. plot this for each eof
fig, ax = plt.subplots(5)
for i in range(5):
    ax[i].plot(time[0], V[i])
    ax[i].set_xlabel('time (1992 year days)')
    ax[i].set_ylabel('Beta') #expansion coefficient
    ax[i].set_title(f'EOF{i + 1}')
```

```python
#1d
y = v #the y is set to the v vectors of the data
y = v - v.mean(axis = 1, keepdims = True) #subtract the mean

U, A, V = np.linalg.svd(y, full_matrices = False) #do the svd
b = U.T @ y

"""
I use EOF1 and EOF2 to distinguish signal from noise. 
As shown in the 1a, EOF 1 explains ~94% of the variance and EOF2 explains ~5%.
I would consider these significant because they each explain a significant portion of the data. 
Together, they explain ~99% of the variance in the data. 
Beyond EOF1 and EOF2, very little of the variance can be further explained by the EOFs(each is less than 1%)
For this reason, I use EOF1 and EOF2 to recreate the data below.
"""

y2 = U[:,0:2] @ b[:2] #use EOF1 and EOF2 to distinguish the data from the noise

#draw a plot (of the first station) to compare the two
plt.plot(time[0], y[0], label = 'Original Data')
plt.plot(time[0], y2[0], label = 'EOF Fitered Data')
plt.xlabel('time (1992 year days)')
plt.ylabel('v (cm/s)')
plt.legend()

```

```python
#1e
#constants
rho = 1.3
c_d = 1.4*10**-3

#subtract the mean
u_wind = u_wind - np.mean(u_wind)
v_wind = v_wind - np.mean(v_wind)

#calculate tau
tau_x = rho * c_d * np.sqrt(u_wind**2 + v_wind**2) * u_wind
tau_y = rho * c_d * np.sqrt(u_wind**2 + v_wind**2) * v_wind
tau = np.squeeze(np.array([tau_x, tau_y]), axis = 1)

#do the svd of tau
U_tau, A_tau, V_tau = np.linalg.svd(tau)
b_tau = U_tau @ tau

#do the svd of the meridional wind vector
U_v_wind, A_v_wind, V_v_wind = np.linalg.svd(v_wind)
b_v_wind = U_v_wind @ v_wind

#plot the first EOF of each (standardized so they can be compared)
fig, ax = plt.subplots()
ax.plot(time[0], b_tau[0] / np.sum(np.abs(b_tau)))
ax.plot(time[0], -b_v_wind[0] / np.sum(np.abs(b_v_wind)))
ax.set_xlabel('time (1992 year days)')
ax.set_ylabel('Beta (normalized)')

"""
We plot the first (normalized)EOF of each timeseries.They are normalized so they can be plotted on top of each other and easily compared.
We can see that the two timeseries EOFs are very strongly correlated. 
This is expected as the stress is directly dependent on the wind speed (and thereby the v component of windspeed).
"""

```

```python
#2a
y = np.array([[12,14]]).T
std = np.array([[1,1]]).T
mean = np.mean(y)

#get a list of values for the function
start_value = 8
end_value = 18
num_values = 1000 #get a lot of values because we will use this to determine confidence limits
values = np.linspace(8, 18, num_values) 

#calculate the product and add the normalization coefficient to get the likelihood function
prod = np.prod(np.exp(-(y - values)**2 / (2 * std**2)), axis = 0)
norm = 1 / np.trapz(prod, values)
likelihood = norm * prod

#calculate the maximum likelihood (var2*y1 + var1*y2) / (var1 + var2)
max_likelihood = ((np.flip(std)**2).T @ y) / (np.sum(std**2))

#caluclate the standard error of the solution and get the points necessary for plotting
std_error = np.sqrt(np.prod(std**2) / np.sum(std**2))
std_min = (max_likelihood - std_error).item()
std_max = (max_likelihood + std_error).item()
std_min_value = likelihood[np.abs(values - std_min).argmin()]
std_max_value = likelihood[np.abs(values - std_max).argmin()]

#get the cdf to determine the 95% confidence limit
likelihood_cdf = np.cumsum(likelihood) * ((end_value - start_value) / num_values)
ci_95_min = values[np.max(np.where(likelihood_cdf < .025))]
ci_95_max = values[np.min(np.where(likelihood_cdf > .975))]
ci_95_min_value = likelihood[np.max(np.where(likelihood_cdf < .025))]
ci_95_max_value = likelihood[np.min(np.where(likelihood_cdf > .975))]

#plot the likelihood function
fig, ax = plt.subplots()
ax.plot(values, likelihood)
ax.set_xlabel('temp (C)')
ax.set_ylabel('likelihood')

#plot the bounds of the confidence interval on the plot
ax.plot([std_min, std_min], [0, std_min_value], c = 'orange')
ax.plot([std_max, std_max], [0, std_max_value], c = 'orange')
ax.plot([ci_95_min, ci_95_min], [0, ci_95_min_value], c = 'red')
ax.plot([ci_95_max, ci_95_max], [0, ci_95_max_value], c = 'red')

print(f'maximum likelihood solution: {max_likelihood.item()}C')
print(f'standard error: +-{round(std_error, 2)}C (in orange)')
print(f'95% confidence limit upper bound: {round(ci_95_max, 2)}C (in red)')
print(f'95% confidence limit lower bound: {round(ci_95_min, 2)}C (in red)')

#plot the pdfs of each observation (useful for visual confirmation)
#sample0 = (1 / (std[0] * np.sqrt(2 * np.pi))) * np.exp((-1 / 2) * ((values - y[0]) / std[0])**2)
#sample1 = (1 / (std[1] * np.sqrt(2 * np.pi))) * np.exp((-1 / 2) * ((values - y[1]) / std[1])**2)
#plt.plot(values, sample0)
#plt.plot(values, sample1)
```

```python
#2b
df = 1 #specify the degrees of freedom
max_likelihood = 13 #from previous problem
y = np.array([12,14])
std = np.array([1,1])

#determine j
j = np.sum((max_likelihood - y)**2 / std**2)

#get the cdf of the chi2 distribution
chi2_values = np.linspace(0, 20, 10000) #values of chi2
chi2_cdf = scipy.stats.chi2.cdf(chi2_values, df = df)

#get the max and min values of chi2 necessary to reject the null hypothesis
min_value = chi2_values[np.max(np.where(chi2_cdf < .025))]
max_value = chi2_values[np.min(np.where(chi2_cdf > .975))]

print(f'J: {j}')
print(f'minimum chi2 value: {round(min_value, 2)}')
print(f'maximum chi2 value: {round(max_value, 2)}')


"""
Because the J value is between the minimum and maximum chi2 values allowed by the 95% confidence interval, 
this passes the goodness of fit test. 
Therefore, we do not reject the null hypothesis that two observations are consistent.
"""

#calculate the p-value by starting at a small value and stepping forward until j is no longer within the acceptable bounds of chi2
alpha = .01
while True:
    min_value = chi2_values[np.max(np.where(chi2_cdf <= alpha))]
    max_value = chi2_values[np.min(np.where(chi2_cdf >= 1 - alpha))]
    if j < min_value or j > max_value:
        break
    alpha += .01
p = 2 * alpha
print(f'p-value: {p}')
```

```python
#2c
#rework of part a
y = np.array([[12,14]]).T
std = np.array([[.5,.5]]).T
mean = np.mean(y)

#get a list of values for the function
start_value = 8
end_value = 18
num_values = 1000 #get a lot of values because we will use this to determine confidence limits
values = np.linspace(8, 18, num_values) 

#calculate the product and add the normalization coefficient to get the likelihood function
prod = np.prod(np.exp(-(y - values)**2 / (2 * std**2)), axis = 0)
norm = 1 / np.trapz(prod, values)
likelihood = norm * prod

#calculate the maximum likelihood (var2*y1 + var1*y2) / (var1 + var2)
max_likelihood = ((np.flip(std)**2).T @ y) / (np.sum(std**2))

#caluclate the standard error of the solution
std_error = np.sqrt(np.prod(std**2) / np.sum(std**2))
std_min = (max_likelihood - std_error).item()
std_max = (max_likelihood + std_error).item()
std_min_value = likelihood[np.abs(values - std_min).argmin()]
std_max_value = likelihood[np.abs(values - std_max).argmin()]

#get the cdf to determine the 95% confidence limit
likelihood_cdf = np.cumsum(likelihood) * ((end_value - start_value) / num_values)
ci_95_min = values[np.max(np.where(likelihood_cdf < .025))]
ci_95_max = values[np.min(np.where(likelihood_cdf > .975))]
ci_95_min_value = likelihood[np.max(np.where(likelihood_cdf < .025))]
ci_95_max_value = likelihood[np.min(np.where(likelihood_cdf > .975))]

#plot the likelihood function
fig, ax = plt.subplots()
ax.plot(values, likelihood)
ax.set_xlabel('temp (C)')
ax.set_ylabel('likelihood')

#plot the bounds of the confidence interval on the plot
ax.plot([std_min, std_min], [0, std_min_value], c = 'orange')
ax.plot([std_max, std_max], [0, std_max_value], c = 'orange')
ax.plot([ci_95_min, ci_95_min], [0, ci_95_min_value], c = 'red')
ax.plot([ci_95_max, ci_95_max], [0, ci_95_max_value], c = 'red')

print(f'maximum likelihood solution: {max_likelihood.item()}C')
print(f'standard error: +-{round(std_error, 2)}C (in orange)')
print(f'95% confidence limit upper bound: {round(ci_95_max, 2)}C (in red)')
print(f'95% confidence limit lower bound: {round(ci_95_min, 2)}C (in red)')

#plot the pdfs of each observation (useful for visual confirmation)
#sample0 = (1 / (std[0] * np.sqrt(2 * np.pi))) * np.exp((-1 / 2) * ((values - y[0]) / std[0])**2)
#sample1 = (1 / (std[1] * np.sqrt(2 * np.pi))) * np.exp((-1 / 2) * ((values - y[1]) / std[1])**2)
#plt.plot(values, sample0)
#plt.plot(values, sample1)


#rework of part b
df = 1 #specify the degrees of freedom
max_likelihood = 13 #from previous problem
y = np.array([12,14])
std = np.array([.5,.5]) #because the error bars are two standard deviations, the standard error is 1/2 * error 

#determine j
j = np.sum((max_likelihood - y)**2 / std**2)

#get the cdf of the chi2 distribution
chi2_values = np.linspace(0, 20, 10000) #values of chi2
chi2_cdf = scipy.stats.chi2.cdf(chi2_values, df = df)

#get the max and min values of chi2 necessary to reject the null hypothesis
min_value = chi2_values[np.max(np.where(chi2_cdf < .025))]
max_value = chi2_values[np.min(np.where(chi2_cdf > .975))]

print(f'J: {j}')
print(f'minimum chi2 value: {round(min_value, 2)}')
print(f'maximum chi2 value: {round(max_value, 2)}')


"""
Because the J value is outside the minimum and maximum chi2 values allowed by the 95% confidence interval, 
this fails the goodness of fit test. 
Therefore, we do reject the null hypothesis that two observations are consistent.
"""

alpha = .01
while True:
    min_value = chi2_values[np.max(np.where(chi2_cdf <= alpha))]
    max_value = chi2_values[np.min(np.where(chi2_cdf >= 1 - alpha))]
    if j < min_value or j > max_value:
        break
    alpha += .01
p = 2 * alpha
print(f'p-value: {p}')

"""
We can see that the p-value decreases significantly with the changed error bars.
Because the standard deviation of each measurement is just half that of the previous problem, 
we would expect the likelihood function have much tighter bounds.
This also results in a much lower p-value because the probability of obtaining our measured values given a maximum likelihood of 13
with smaller standard errors, is much lower than with larger standard errors.
We can also see that we reject the null hypoythesis as explained above.
"""
```

```python
#3b
#calculate the mean variance and standard deviation
mean = np.mean(np_sl)
var = np.var(np_sl)
std = np.sqrt(var)

#calculate the pdf from the mean and standard deviation
values = np.linspace(-40, 40, 1000)
pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp((-1 / 2) * ((values - mean) / std)**2)

fig, ax = plt.subplots()
ax.hist(np_sl[0], bins = 20, density = True) #plot the normalized histogram
ax.plot(values, pdf) #plot the pdf
ax.set_xlabel('sea level (cm)')
ax.set_ylabel('probability')

"""
Comparing the normal distribution and the normalized histogram, 
we can see the the pdf roughly matches the observations. 
Although the pdf does not match the observations exactly, 
this is still likely a good fit for the data.
Any significant variance (such as the underestimation of the pdf near the mean), 
is likely a result of the fact that we have so few observations which results in a more noisy histogram. 
"""
```
