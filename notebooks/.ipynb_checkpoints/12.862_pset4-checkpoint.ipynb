## 2. North River estuary
### a. tidal channel

```python
# known variables
a_0 = 2 # tidal amplitude [m]
b = 50 # average width of channel <b> = w [m]
C_d = 3*10**-3 # bottom drag [unitless]
g = 9.8 # accel due to gravity [m/s^2]
h = 2.5 # average depth of the channel <h> = H [m]
U = 1 # water velocity [m/s]
t_m2 = np.arange(0,12.42 * 3600,1) # time [s] 12.42 hrs in a M2 tide --> convert to seconds
omega = 1.4*10**-4 # frequency [rad/s]
x_head = 30000 # distance along the channel to the end [m]
x_mouth = 0 # [m]
x_ten = 10000

# variables needed to be calculated from knowns
r = 0.85*((C_d*U)/h)
c = np.sqrt((2*omega*g*h*b)/(r*b))
k = omega/c
U_0 = (a_0/h)*(np.sqrt((omega*g*h*b)/(r*b)))

# unknowns 
# eta = a_0*(np.exp(-k*x))*np.cos((omega*t_m2)-(k*x)) # representative eqn for eta
# u = U_0*(np.exp(-k*x))*np.cos((omega*t_m2)-(k*x) + (np.pi/4))
```

```python
# tidal amplitude
eta_mouth = []
eta_head = []
eta_ten = []
for i in range(len(t_m2)):
    eta_mouth.append(a_0*(np.exp(-k*x_mouth))*np.cos((omega*i)-(k*x_mouth)))
    eta_head.append(a_0*(np.exp(-k*x_head))*np.cos((omega*i)-(k*x_head)))
    eta_ten.append(a_0*(np.exp(-k*x_ten))*np.cos((omega*i)-(k*x_ten)))
```

```python
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Tidal Amplitude Throughout the Tidal Cycle', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(t_m2,eta_ten,label='x = 10 km')
ax.plot(t_m2,eta_mouth,label='Mouth')
ax.plot(t_m2,eta_head,label='Head')
ax.plot(t_m2,np.subtract(eta_head,eta_mouth),label='Head – Mouth')



# axes formatting
ax.set_xlabel('Time [s]', fontweight='bold')
ax.set_ylabel('\u03B7 [m]',fontweight='bold')
ax.legend();
ax.set_ylim([-3,3])
ax.set_xlim([0,12.42 * 3600])
ax.vlines([(12.42 * 3600)/2],-3,3,color='k',linestyle='--', alpha=0.1);
```

```python
# tidal velocity
u_mouth = []
u_head = []
u_ten = []
for i in range(len(t_m2)):
    u_mouth.append(U_0*((np.exp(-k*x_mouth))*np.cos((omega*i)-(k*x_mouth)+(np.pi/4))))
    u_head.append(U_0*((np.exp(-k*x_head))*np.cos((omega*i)-(k*x_head) + (np.pi/4))) )
    u_ten.append(U_0*((np.exp(-k*x_ten))*np.cos((omega*i)-(k*x_ten) + (np.pi/4))))
```

```python
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Tidal Velocity Throughout the Tidal Cycle', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(t_m2,u_ten,label='x = 10 km')
ax.plot(t_m2,u_mouth,label='Mouth')
ax.plot(t_m2,u_head,label='Head')
ax.plot(t_m2,np.subtract(u_head,u_mouth),label='Head – Mouth')



# axes formatting
ax.set_xlabel('Time [s]', fontweight='bold')
ax.set_ylabel('u [m/s]',fontweight='bold')
ax.legend();
ax.set_ylim([-2,2])
ax.set_xlim([0,12.42 * 3600])
ax.vlines([(12.42 * 3600)/2],-3,3,color='k',linestyle='--', alpha=0.1);
```

### b. salt marshes on either side of the tidal channel

```python
b = 250 # when salt marshes are inundated 
w = 50
c = np.sqrt((2*omega*g*h*b)/(r*w))
k_b = omega/c

b = 50 # when salt marshes are *not* inundated
c = np.sqrt((2*omega*g*h*b)/(r*b))
k = omega/c

# redo tidal amplitude for salt marsh case

# tidal amplitude
eta_mouth_salt = []
eta_head_salt = []
eta_ten_salt = []
for i in range(len(t_m2)):
    if (i<=((12.42*3600)/2)):
        eta_mouth_salt.append(a_0*(np.exp(-k*x_mouth))*np.cos((omega*i)-(k*x_mouth)))
        eta_head_salt.append(a_0*(np.exp(-k*x_head))*np.cos((omega*i)-(k*x_head)))
        eta_ten_salt.append(a_0*(np.exp(-k*x_ten))*np.cos((omega*i)-(k*x_ten)))
        
    else:
        eta_mouth_salt.append(a_0*(np.exp(-k_b*x_mouth))*np.cos((omega*i)-(k_b*x_mouth)))
        eta_head_salt.append(a_0*(np.exp(-k_b*x_head))*np.cos((omega*i)-(k_b*x_head)))
        eta_ten_salt.append(a_0*(np.exp(-k_b*x_ten))*np.cos((omega*i)-(k_b*x_ten)))
```

```python
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Tidal Amplitude Throughout the Tidal Cycle', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(t_m2,eta_ten_salt,label='x = 10 km')
ax.plot(t_m2,eta_mouth_salt,label='Mouth')
ax.plot(t_m2,eta_head_salt,label='Head')
ax.plot(t_m2,np.subtract(eta_head_salt,eta_mouth_salt),label='Head – Mouth')

# axes formatting
ax.set_xlabel('Time [s]', fontweight='bold')
ax.set_ylabel('\u03B7 [m]',fontweight='bold')
ax.legend();
ax.set_ylim([-3,3])
ax.set_xlim([0,12.42 * 3600])
ax.vlines([(12.42 * 3600)/2],-3,3,color='k',linestyle='--', alpha=0.1);
```

```python
# tidal velocity
u_mouth_salt = []
u_head_salt = []
u_ten_salt = []
for i in range(len(t_m2)):
    if (i<=((12.42*3600)/2)):
        u_mouth_salt.append(U_0*((np.exp(-k*x_mouth))*np.cos((omega*i)-(k*x_mouth)+(np.pi/4))))
        u_head_salt.append(U_0*((np.exp(-k*x_head))*np.cos((omega*i)-(k*x_head) + (np.pi/4))) )
        u_ten_salt.append(U_0*((np.exp(-k*x_ten))*np.cos((omega*i)-(k*x_ten) + (np.pi/4))))

    else:
        u_mouth_salt.append(U_0*((np.exp(-k_b*x_mouth))*np.cos((omega*i)-(k_b*x_mouth)+(np.pi/4))))
        u_head_salt.append(U_0*((np.exp(-k_b*x_head))*np.cos((omega*i)-(k_b*x_head) + (np.pi/4))) )
        u_ten_salt.append(U_0*((np.exp(-k_b*x_ten))*np.cos((omega*i)-(k_b*x_ten) + (np.pi/4))))
```

```python
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Tidal Velocity Throughout the Tidal Cycle', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(t_m2,u_ten_salt,label='x = 10 km')
ax.plot(t_m2,u_mouth_salt,label='Mouth')
ax.plot(t_m2,u_head_salt,label='Head')
ax.plot(t_m2,np.subtract(u_head,u_mouth),label='Head – Mouth')



# axes formatting
ax.set_xlabel('Time [s]', fontweight='bold')
ax.set_ylabel('u [m/s]',fontweight='bold')
ax.legend();
ax.set_ylim([-2,2])
ax.set_xlim([0,12.42 * 3600])

ax.vlines([(12.42 * 3600)/2],-3,3,color='k',linestyle='--', alpha=0.1);
```

### c. estimate estuarine circulation

```python
# average tidal velocity
x=0 # at the mouth [m]
u = U_0*(np.exp(-k*x))*np.cos((omega*t_m2)-(k*x) + (np.pi/4)) # velocities at the mouth [m/s]
U_T = np.mean(u) # tidal velocity [m/s]
von_k = 0.41 # von karman constant [unitless]

K_m = von_k*U_T*h # eddy viscosity [m^2/s]

del_s = 30-1 # difference between salinities [PSU]
del_x = 15000 # length of salinity gradient [m]
sal_grad = del_s/del_x # [PSU/m]

beta = 7.7*10**-4

u_e = (1/48)*((beta*sal_grad*(h**3))/(K_m))
u_e
```

```python
s_prime = ((sal_grad**2)*(h**3))/(U_T**2)
s_prime
```

```python
A = h*b

flux_salt = u_e*A*del_s
flux_salt
```
