## Problem 1. Model of exponential decay.

Consider observations of a decaying temperature anomaly,

$y(t_1=1 yr)= 5.0 ± 0.3 K$

$y(t_2=2 yr)= 2.0 ± 0.3 K$, and

$y(t_3=3 yr)= 1.0 ± 0.3 K$.

We have theoretical reasons to believe that the signal is decaying exponentially according to the model:

$Θ(t) = a e^{-t}$,

where $Θ(t)$ is temperature and “a” is an unknown constant.

### a. 
#### Solve for the minimum-variance estimate of “a” and its standard error. Also find the temperature at $t=0$ years, i.e., $Θ(t₀=0 yr)$.

There are 2 ways to solve this problem. The first is to use the minimum-variance estimate. I will use the equation from page 140 of the notes that says:
    
$\tilde u = −(\Gamma^TW^{−1}\Gamma)^{−1}\Gamma^TW^{−1}y$

Where $\tilde u = \tilde x_{MV}$ and $W$ is a weighting matrix we'll set $W=I$ and $\Gamma$ is our model where $\Gamma = \begin{bmatrix}
e^{-t_1}\\
e^{-t_2}\\
e^{-t_3}
\end{bmatrix}$

```python
'DEFINE THE VARIABLES'
y = np.array([np.array([5,2,1])]).T
E = np.array([[np.exp(-1), np.exp(-2), np.exp(-3)]]).T
n = np.array([0.3,0.3,0.3])

W = np.eye(3) # weighting matrix, let it be idenity matrix since all obs are equally trusted
Gamma = E # for MV estimation

'FIRST METHOD: MINIMUM VARIANCE ESTIMATION'
x_MV = ((-(Gamma.T@np.linalg.inv(W)@Gamma))**(-1))@(Gamma.T@np.linalg.inv(W)@y)

print('x_ML = '+str(abs(round(x_MV[0][0],2))))
```

The second way is to use maximum likelihood estimation. In the notes on page 117 it says "the minimum variance estimate [...] coincides with the maximum likelihood estimate when error statistics are
normally-distributed". Thus,

$\tilde x_{MV} = \tilde x_{ML}$.

Therefore we can use the maximum likelihood estimation, because our errors are normally distributed, where 

$x_{ML} = (E^TC_{nn}^{-1}E)^{-1}E^TC_{nn}^{-1}y$

```python
'DEFINE THE VARIABLES'
C_nn = (np.diag(n))**2

'SECOND METHOD: MAXIMUM LIKELIHOOD ESTIMATION'
x_ML = (np.linalg.inv(E.T@np.linalg.inv(C_nn)@E))@E.T@np.linalg.inv(C_nn)@y

print('x_ML = '+str(round(x_ML[0][0],2)))
```

Thus, I have shown that either method gets the same answer for $a = x_{MV} = x_{ML} = 13.83$. 

To get the uncertainity (standard error) for this solution use 

$P = (E^TC_{nn}^{−1}E)^{−1}$

```python
'FIND UNCERTAINITY'
P = np.sqrt(np.linalg.inv(E.T@np.linalg.inv(C_nn)@(E)))

print('Standard error of our solution estimate is '+str(round(P[0][0],2))+'.')
```

Using the model we've now constructed we can find an estimate for time t=0 years.

$y = e^{t=0 years}\cdot x_{MV} \pm \sigma_{\tilde y}$ where $\sigma_{\tilde y} = \sigma_{\tilde x}e^{-t}$ 

because the error will grow exponentially backwards in time (where we don't have as much information).

```python
y_at_t_0 = np.exp(0)*x_MV
sigma_x_tilde = np.sqrt(P)
sigma_y_tilde = sigma_x_tilde*np.exp(0)

print('The temperature at t=0 years is = '+str(round(-y_at_t_0[0][0],2))+' +- '+str(round(sigma_y_tilde[0][0],2))+' deg C')
```

##### First, we will perform a $\chi^2$ goodness-of-fit test.

Our null hypothesis is that the observations are consistent with the exponential model.

To 

• choose $\alpha$, the level of significance, and find the region of acceptance defined by the upper and lower bounds, $\chi_U$ and $\chi_L$, where $P(x_L) = \alpha$ and $P(x_U) = 1-\alpha$ and $P$ is the expected cumulative distribution function of the test statistic

• vary $\alpha$ until the sampled test statistic, $\textbf{J}$ has a value at the edge of the region of acceptance

• the probability that the hypothesis is wrongly rejected is the p-value and is equal to $2\alpha$

Our null hypothesis is that the two observations $y_1, y_2$ are consistent. First, we will state that our test statistic $J$:

$J = \frac{(y_1 -x_{true})^2}{\sigma_1^2} + \frac{(y_2 -x_{true})^2}{\sigma_2^2} + \frac{(y_3 -x_{true})^2}{\sigma_3^2}$

Second, we will compute the value of $\tilde J$.

## Problem 2. Combining a model of vertical structure with observations.

This problem is inspired by the vertical structure of the North Pacific Ocean as illustrated in the original Abyssal Recipes paper (Munk 1966). Consider a tracer, such as radiocarbon, that has a steady-state vertical profile, $C(z)$, due to the competing effects of turbulent vertical mixing and radioactive decay. The relevant differential equation is:

$\begin{equation}
\frac{d^2 C}{dz^2} - \kappa^2 C = 0,
\end{equation}$


where $\kappa^2 = 1.0 km^{-2}$. The goal of this problem is to solve for the vertical profile at a number of depths, $C(-nΔz)$, where $Δz=0.1 km$ and $n$ goes from $0$ to $50$. (Note: $z$ points upward in the ocean by convention, so our axis will start at zero at the surface and then use negative $z$ values when descending into the ocean. When $C$ represents radiocarbon, it has units of "fraction modern" which is dimensionless.)

### a) 
#### Translate the differential equation into a set of simultaneous equations at the depths, $C(-nΔz)$, for $n=1$ to $49$. Put the set of equations into matrix form, $𝐀𝐱 = 𝐛$, where all variables are defined.

Following along with equations from page 132 of the notes (where models are combined with observations) we can construct matrices $A, x, b$.

$A$ is the model matrix, $x$ is the set of physical variables, and $b$ are the model constraint or forcing. 

$ A = \begin{bmatrix}
1/(\Delta z^2) & -2/(\Delta z^2) -\kappa^2 & 1/(\Delta z^2) & 0 & ... & 0 \\
0  & 1/(\Delta z^2) & -2/(\Delta z^2) -\kappa^2 & 1/(\Delta z^2) & 0 & ...\\
... & ... & ... & ... & ... & 0 \\
0 & ... & 0 & 1/(\Delta z^2) & -2/(\Delta z^2) -\kappa^2 & 1/(\Delta z^2)
\end{bmatrix}$, $ x = \begin{bmatrix}
x_1\\
x_2\\
...\\
x_N
\end{bmatrix}$ , $ b = \begin{bmatrix}
b_1\\
b_2\\
...\\
b_N
\end{bmatrix}$

### b)

 With this one observation we can write:
 
 $Ex + n = y$ such that $E$ is $51x1$.

### c) 
#### Concatenate the equations of the physical and observational models. Find the particular SVD solution for $C(-nΔz)$ for $n = 0$ to $50$ and plot it. What is the rank of the problem?

```python
E.shape
```

```python
'DEFINE THE VARIABLES'
delta_z = 0.1 # km
kappa_squared = 1 # 1/km^2
n = np.arange(0,51,1) # scalar
z = -n*delta_z # km, depth
E = np.zeros((1,51))
E[0,0] = 1 # observation at surface
y = np.ones((1,1))

'COMPUTE A'
A = np.zeros((49,51))# initialize matrix with zeros

np.fill_diagonal(A[:,2:],1/(delta_z**2)) #  upper diagonal
np.fill_diagonal(A[:,1:],(-2/(delta_z**2)) - kappa_squared) #  center diagonal
np.fill_diagonal(A[0:,:],1/(delta_z**2)) #  lower diagonal

'DEFINE b'
b = np.zeros((49,1))

'CONCATENATE A & E AND B & Y'
A_full = np.concatenate((E,A), axis=0)
b_full = np.concatenate((y,b), axis=0)

'COMPUTE x'
x = np.linalg.pinv(A_full)@b_full

'PLOT RESULTS'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
fig.suptitle('Radiocarbon profile for North Pacific Ocean', fontsize=14, y=0.925, fontweight='bold')

# plot observations
ax.plot(x,z, color='blue')

#axes formatting
ax.set_xlabel('Concentration of radiocarbon tracer', fontweight='bold')
ax.set_ylabel('Depth [m]',fontweight='bold');
ax.set_ylim([-5,1])
ax.set_xlim([-0.1,1])
ax.vlines(0,1,-5,color='k',linestyle='--', alpha=0.5);
```

```python
print('rank of A is = '+str(np.linalg.matrix_rank(A_full)))
```

The rank of any problem is between 1 and the higher of the matrix dimensions M,N. The rank is equal to the maximal number of independent columns. In this case the rank of A is 50.

### d) 
#### Find an expression for all other solutions to the problem and plot a handful of them.

All the other solutions to the problem lie in the null space. We have a matrix A and all the vectors B such that AB = 0 but the B vectors $\neq 0$. These vectors make up the null space. The size of the null space tells us about the number of linear relationships between variables, since the null space is a linear combination of vectors that $= 0$. So we can find the null space and multiply it by random vectors and A to get some of the few possibilities of solutions for the problem.

The expression we find for all other solutions is = $\tilde x_p + (V_{null}\cdot n_{rand})$ 

where $n_{rand}$ is a random value scalar (we multiply by a scalar because the rank of the null space is 1 (there's only one linear relationship between the variables) and $\tilde x_p = V \Lambda^{-1} U^{T} y$ is the particular solution.

```python
'DEFINE VARIABLES'
y = b_full

'COMPUTE SVD'
(U,lambdas,Vh)  = np.linalg.svd(A_full, full_matrices=True) 

'DEFINE V, NULL SPACE, LAMBDA'
Vh = Vh.T
Vh_not_null = Vh[:,:50]
Vh_null = Vh[:,50:]
Lambda = np.diag(lambdas)

'COMPUTE PARTICULAR SOLUTION'
x_tilde_p = Vh_not_null@np.linalg.inv(Lambda)@U.T@y

'COMPUTE NULL SPACE SOLUTION'
x_tilde_p_null = x_tilde_p + (Vh_null*np.random.rand(1))

'PLOT 100 RESULTS'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
fig.suptitle('100 Radiocarbon profiles for North Pacific Ocean', fontsize=14, y=0.925, fontweight='bold')

# plot observations
for i in range(100):
    'COMPUTE NULL SPACE SOLUTION'
    x_tilde_p_null = x_tilde_p + (Vh_null*np.random.rand(1))
    ax.plot(x_tilde_p_null,z)

#axes formatting
ax.set_xlabel('Concentration of radiocarbon tracer', fontweight='bold')
ax.set_ylabel('Depth [m]',fontweight='bold');
ax.set_ylim([-5,1])
ax.set_xlim([-0.1,1])
ax.vlines(0,1,-5,color='k',linestyle='--', alpha=0.5);
```

## Problem 3. Combining a model with time-evolving observations.

Consider the observations of Woods Hole daily sea surface temperature anomaly,  $y(t) $, given in the text file  $WoodsHoleSST.txt $. The observations begin at  $t = 0 $ day and the record length is  $100$ $days $. The standard error of each observation is  $0.10°C $. The goal is to explain the observations using the basic equation:

$\begin{equation}
\frac{dΘ}{dt} = ℱ(t)
\end{equation}$


(i.e., the “model”), where ℱ(t) is related to the air-sea heat flux.



### a) 
#### Discretize the model equation with $Δt = 1 day $, such that the discrete representation of  $ℱ(t)$ is a independent Gaussian white noise process of variance of  $0.25 (°C/d)² $ and mean of zero. Write a time-stepping rule for the first-guess model state,  $x₀(t+1) $, in terms of  $x₀(t) $.



We can discretize the model with a forward-time approach in which

$x(t+1)-x(t)=ℱ(t)\Delta t$ 

Our time-stepping rule for the first-guess model state is:

$x_0(t+\Delta t) = x_0(t) + =ℱ(t)\Delta t$

### b) 
#### Write a rule for the uncertainty of the first-guess model state, $<(x₀(t+1)′)²>$, in terms of things known at time t or earlier. We can assume that the model is unbiased.

We can also write the variance of our model as:

$<(x_0'(t+\Delta t))^2> = <(x_0'(t))^2> + <(ℱ'(t))^2>\Delta t^2$

where $<(x_0'(t))^2>$ is the variance of our observations at $t$, $<(ℱ'(t))^2>$ is the variance of the forcing at $t$ times how long we've been going $\Delta t$ therefore $<(x_0'(t+\Delta t))^2>$ is the uncertainity of the first-guess model at $t + \Delta t$.

### c) 
#### Use the first observation as the model initial condition, $x₀(t = 0 d)$. Then use the model to simulate $x₀(t = 1 d)$. What is the expected value of the first-guess model simulation, <x₀( t = 1 d)>? (Don’t use the observation at $t= 1 d$ in this step.)

```python
'DEFINE VARIABLES'
y_err = 0.1
y = WoHoSST
F_err = math.sqrt(0.25)
F_expected = 0 
F = np.random.normal(F_expected, F_err) # single random value of F
delta_t = 1
x_t_0 = WoHoSST[0]

'USE MODEL TO SIMULATE X_0(T=1D)'
x_1_simulated = x_t_0 + F*delta_t
print('x_1_simulated = ',str(round(x_1_simulated,2)),'deg C')

'CALCULATE EXPECTED VALUE OF X'
x_1_expected = x_t_0 + F_expected*delta_t
print('x_1_expected = ',str(round(x_1_expected,2)),'deg C')
```

### d) 
#### What is the uncertainty of the model prediction at $t = 1 d (i.e., <(x₀(1 d)′)²>$?

```python
'CALCULATE UNCERTAINITY OF MODEL PREDICTION'
p_1 = np.sqrt(y_err**2 + (F_err**2 * delta_t**2))
print('uncertainty p_1 = ',str(round(p_1,2)),'deg C')
```

### e) 
#### By continuing to integrate the forward model through time (ignore the additional observations), make a general rule for the uncertainty of the model estimate at any time. Plot the model estimate and its uncertainty through time.

```python
'INITIALIZE LISTS'
x_var = []
x_exp = []

x_t_prev = WoHoSST[0] # solution for one time step before current
var_prev = 0.1**2 # variance for one time step before current 

x_var.append(var_prev)
x_exp.append(x_t_prev)

for i in range(100):

    'CALCULATE VALUES'
    var_t =  var_prev + (F_err**2 * delta_t**2)
    x_t_expected = x_t_prev + F_expected*delta_t
    
    'UPDATE LISTS'
    x_var.append(var_t)
    x_exp.append(x_t_expected)
    
    x_t_prev = x_t_expected
    var_prev = var_t

'PLOT RESULTS'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Model Prediction for WoHo Temps (100 days)', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.scatter(WH_time, WoHoSST,label='Observations')
ax.plot(WH_time, x_exp, label='Expected');
ax.fill_between(WH_time,(x_exp - np.sqrt(x_var)), (x_exp + np.sqrt(x_var)), color='navy', alpha=0.2, label= 'Model Error')


# axes formatting
ax.set_xlabel('Time [day]',fontweight='bold');
ax.set_ylabel('Temperature [C]',fontweight='bold');
ax.legend();
```
