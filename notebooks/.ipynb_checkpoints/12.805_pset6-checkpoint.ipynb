## Problem 1. 

Write a function (computer code) that will produce a one-sided, band-averaged estimate of 
the autospectral densityyou will turn in your code, after checking it in the following 
problems.  Follow the recipe in Section $4.7.1$ of the class notes on spectral analysis.  This should 
be a function that takes as inputs:

* $y$, vector to be transformed (assumed real and uniformly spaced) 
* $dt$, time increment 
* $M$, number of bands to average 
 
It should output the band-averaged spectrum and frequency vector and plot the spectrum on a 
log-log scale, with a $95%$ confidence interval.   
 
* To help, I have provided the following functions (in MATLAB and Python):  
    * centered_FFT.m. Computes FFT, with zero frequency in the center, and returns dimensional frequency vector. 
    * band_avg.m Computes block averages for band averaging 
    * confid.m computes a confidence interval 
    * confidence_interval.m plots confidence interval on a log-log scale 
* I have made the same functions in python (in functions_12805.py). 
* This is meant to be straightforwardif you find it difficult, please contact me and I will 
help you. 

```python
def band_avg_autospec_dens(x, dt, M, taper):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    This function produces a one-sided, band-averaged estimate of the autospectral density of a given 
    time series x, with time increment dt, and number of bands M. 
    
    Input:
        x (array)                   : vector to be transformed (real & uniformly spaced)
        dt (float)                  : time increment
        M (float)                   : number of bands to average 
        taper (bool)                : if True, taper the time series before the FFT

    Output: 
        spec_dens_avg (array)       : band-averaged autospectral density of the time series x
        fm_avg (array)              : band-averaged frequency of the time series x
        lower_ci (float)            : lower confidence limit for a 95% confidence interval
        upper_ci (float)            : upper confidence limit for a 95% confidence interval
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    'DEFINE VARIABLES'
    N = x.size # number of observations
    T = N * dt # total time of observations 
    
    '1. COMPUTE THE MEAN OF THE TIME SERIES x(t) AND SUBTRACT IT FROM THE TIME SERIES'
    x_anom = x - x.mean()
    
    if taper: 
        '2. TAPER THE RECORD USING A TAPER WINDOW xtaper(t) = w(t)x(t)'
        w = scipy.signal.windows.tukey(len(x_anom)) # Tukey window
        xtaper = w*x
    
    '3. COMPUTE THE FFT OF THE ENTIRE TAPERED RECORD'
    x_fft, fm = centeredFFT(x_anom, dt)
    
    '4. GENERATE A FREQUENCY VECTOR fm = m/(Nt) WITH m RANGING FROM N/2 TO N/2  1 (FOR EVEN N)'
    # generated from the centeredFFT function, fm
    
    '5. COMPUTE THE RAW, ONE-SIDED SPECTRAL DENSITY AND DISCARDED FREQUENCIES  0'
    # equation 4.27 from the notes, psi_k = autospectral density
    spec_dens = ((2*T)/(N**2))*(x_fft*(np.conj(x_fft)))
    
    # only keep positive frequencies, negative freq don't have physical meaning
    spec_dens = spec_dens[np.where(fm > 0)]
    fm = fm[np.where(fm > 0)]
        
    '6. BAND AVERAGE THE RAW SPECTRUM OVER nd FREQUENCY BANDS'
    spec_dens_avg = band_avg(spec_dens, M)
    fm_avg = band_avg(fm, M)
        
    '7. ESTIMATE A CONFIDENCE INTERVAL (TYPICALLY A 95% CONFIDENCE INTERVAL)'
    lower_ci, upper_ci = confid(0.05,2*M)
    
    return spec_dens_avg, fm_avg, lower_ci, upper_ci
```

```python
def band_avg_autospec_dens_plot(fm_avg, spec_dens_avg, M, title, fig_quality, lineColor, confidColor, ifColor):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    This function produces a plot of the autospectral density of a time sereis spec_dens_avg plotted against
    the frequency averaged fm_avg with a 95% confidence interval.
    
    Input:
        spec_dens_avg (array)       : band-averaged autospectral density of the time series x
        fm_avg (array)              : band-averaged frequency of the time series x
        title (string)              : title of the figure 
        fig_quality (int)           : quality of the figure, e.g. 100 dpi
        lineColor (string)          : color of the line plot
        confidColor (string)        : color of the confidence interval
        ifColor (bool)              : if True use custom colors, else False use default color scheme

    Output: 
    * returns plot of the band-averaged autospectral density of a time-series plotted against the frequencies
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    # create figure
    fig,ax = plt.subplots(figsize=(10,7))
    fig.set_dpi(fig_quality)
    fig.suptitle(title, fontsize=14, y=0.925, fontweight='bold')

    # plot
    if ifColor:
        ax.loglog(fm_avg, spec_dens_avg, color=lineColor, alpha=0.8);
        confidence_interval(0.05,M*2,confidColor)
    else:
        ax.loglog(fm_avg, spec_dens_avg, color='#2c7fb8', alpha=0.8);
        confidence_interval(0.05,M*2,'#253494')

    # axes formatting
    ax.set_xlabel('Frequency [1/s]',fontweight='bold');
    ax.set_ylabel('Autospectral Density [(m/s)^/(1/s)]',fontweight='bold');
    
    return fig, ax
```

## Problem 4. 

Low-pass filter the zonal velocity, $U$, using a 1-day (24-point) running average.
1.  First, construct a rectangular filter window (a vector of ones) and apply an appropriate normalization (see Eqns $4.33-4.35$ and $4.42$ of the class notes).  
2. Then, form the filtered time series by convolution. 
3. Plot $U$ and $Ubar$.  
4. Plot the spectra of $U$ and $Ubar$, and the ratio of these spectra (plot the spectrum of Ubar over the spectrum of U).  This is one way to estimate the filter transfer function of your running average filter.
5. Estimate the half-power point of the 1-day running average filter using your graph. 
 
Hints:  
* MATLAB: $Ubar$=conv($U$,window,same) 
* Python: something like $Ubar$ = np.convolve($U$, window, mode='same') 
* Try using about 11 bands for averaging 
* Try plotting the ratio with linear axes when estimating the half-power point 

<font color='blue'> The convolution theorem says, "the Fourier transform of a convolution of two variables is the product of their Fourier transforms conversely, the Fourier transform of a product of two variables is the convolution of
their Fourier transforms". </font>  


```python
'DEFINE VARIABLES'
U = sc_U # for simplicity 

'1. CONSTRUCT RECTANGULAR FILTER WINDOW AND APPLY APPROPRIATE NORMALIZATION'
run_avg_len = 24
rec_filt_window = np.ones(run_avg_len)/run_avg_len # normalize by dividing by 24

'2. FORM THE FILTERTED TIME SERIES BY CONVOLUTION'
Ubar = np.convolve(np.squeeze(U), rec_filt_window, mode='same')

'3. PLOT U AND UBAR'
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
fig.suptitle('U vs. Ubar Time Series', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(U, color='#2c7fb8', alpha=0.3, label='U');
ax.plot(Ubar, color='#2c7fb8', alpha=1, label='Ubar');

# axes formatting
ax.set_xlabel('Time [day]',fontweight='bold');
ax.set_ylabel('Velocity [m/s]',fontweight='bold');
ax.legend();

'4. PLOT THE SPECTRA OF U, UBAR AND PLOT THE RATIO OF THESE SPECTRA'

'SPECTRA PLOT'
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
fig.suptitle('Band-averaged autospectral density of U & Ubar', fontsize=14, y=0.925, fontweight='bold')

# plot U
ax.loglog(fm_avg_U, spec_dens_avg_U, color='#de77ae', alpha=0.8, label='U'); 
confidence_interval(0.05,M*2,'#c51b7d')

# plot Ubar
ax.loglog(fm_avg_Ubar, spec_dens_avg_Ubar, color='#7fbc41', alpha=0.8, label='Ubar'); 
confidence_interval(0.05,M*2,'#4d9221')

# axes formatting
ax.set_xlabel('Frequency [1/s]',fontweight='bold');
ax.set_ylabel('Autospectral Density [(m/s)^/(1/s)]',fontweight='bold');
ax.legend();
    

'RATIO PLOT'
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(75)
fig.suptitle('Ratio of U vs. UBar Spectrum', fontsize=14, y=0.925, fontweight='bold')

# plot ratio
ratio = spec_dens_avg_Ubar/spec_dens_avg_U
ax.plot(ratio, color='purple', alpha=1);
ax.set_xscale('log')

# axes formatting
ax.set_xlabel('Frequency [1/s]',fontweight='bold');
ax.set_ylabel('Spectra Ratio [Ubar/U]',fontweight='bold');
ax.scatter(np.where((ratio>0.49) & (ratio<0.50)), ratio[np.where((ratio>0.49) & (ratio<0.50))].real[0], color='purple')
ax.vlines(np.where((ratio>0.49) & (ratio<0.50)), 0, 1, color='k',linestyle='--', alpha=0.5);
ax.text(22, 0.5, 'Half-power point = '+str(fm_avg_U[np.where((ratio>0.49) & (ratio<0.50))]));


'5. ESTIMATE THE HALF-POWER POINT OF THE 1-DAY RUNNING AVERAGE FILTER'
half_power_point = np.where((ratio>0.49) & (ratio<0.50))
f_c = fm_avg_U[half_power_point]
print('Half-power point frequency = '+str(f_c))
```
