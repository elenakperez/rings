## Problem 1. Mean of an autocorrelated timeseries (14 points)

Download the text file containing the Atlantic meridional overturning circulation (MOC) at $26.5°N$, in monthly averages starting with April, 2004 (moc24N_monavg.txt). The standard error in the inference of monthly mean MOC is $0.1 Sv$ with an expected normal distribution.


### a. 
#### Diagnose the maximum likelihood solution for the mean MOC and its standard error under the assumption that each monthly average is independent and the variance of the MOC is $100 Sv^2$. You may have to derive a solution formula that is unique from any found in the class notes. (5 points)

From the problem statement we know that the standard error $\sigma_n = 0.1 Sv$, that is the error on each observation slash the square root of the solution covariance of the noise, with an expected normal distribution. We also know the variance of the MOC as a system is $\sigma_x^2 = 100 Sv$. 

We want to find the max likelihood solution for the mean MOC $\tilde x_{ML}$ and its standard error $\sigma_{\tilde x_{ML}}$.

In order to find the ML solution we start with the model $y = E\bar x + x' + n$ where $\bar x$ is the mean $x$, $x'$ are the fluctuations around the mean $\bar x$, and $n$ is the noise. Define $q = x' + n$, the combination of observational and other noise, so we can rewrite our model as $y = E\bar x + q$. 

From the notes it says "If the noise statistics are normal, the most convenient way to find the maximum-likelihood
solution is to solve a least-squares problem where we choose the weighting matrix, $W= C_{nn}$". For us this is a new matrix called $C_{qq}$.

The covariance matrix for our combined observational and noise is defined as $C_{qq} = <(x' + n)><(x' + n)> = C_{x'x'} + C_{nn}$. Thus $C_{qq}$ can also be written as a matrix with $\sigma_x'^2 + \sigma_n^2$ on the diagonals.

In order to find the ML solution we want to minimize the cost function $J = q^T C_{qq}^{-1} q$. 

This has become a weighted least squares problem where our weighting matrix is $C_{qq}$.

Therefore our ML solution $\tilde x_{ML}$ is given by equation 2.79 in the notes $\tilde x_{ML} = (E^T C_{qq}^{-1} E)^{-1}E^T C_{qq}^{-1}$ where we've replaced $C_{nn}$ with $C_{qq}$ since is our weighting matrix.

Our uncertainity is given by equation 2.82 in the notes $P = (E^T C_{qq}E)^{-1}$ where, again, we've replaced $C_{nn}$ with $C_{qq}$ since is our weighting matrix. Thus, $\sigma_{\tilde x_{ML}} = (E^T C_{qq}E)^{-1}$

The rest of our variables are the matrix $E$ is a column vector of 1s since our background state is not varying, $y$ is our MOC timeseries (our observations), and $q$ is a combination of fluctuations/noise. 

Now we have all the equations and components needed to find the max likelihood solution for the mean MOC $\tilde x_{ML}$ and its standard error $\sigma_{\tilde x_{ML}}$.



```python
'CREATE VARIABLES'
y = np.array([moc24_monavg]).T # observations Nx1 (N = 173 in this case)
E = np.ones((y.size, 1)) # observational matrix Nx1
sigma_n = 0.1 # solution covariance of the noise 
sigma_x = 10 # variance of the system 
q = (sigma_n**2) + (sigma_x**2) # observation + random noise 

'COMPUTE C_QQ MATRIX'
C_qq = np.diag(np.repeat(q, y.size)) # initialize observational & noise covariance matrix with dimensions NxN

'USE MODIFIED EQN 2.79 TO FIND X_ML'
x_ml = (np.linalg.inv(((E.T@(np.linalg.inv(C_qq))@E))))@((E.T@(np.linalg.inv(C_qq)))@y)

'USE MODIFIED EQN 2.82 TO FIND UNCERTAINITY SIGMA_X_ML'
sigma_x_ml = np.sqrt(np.linalg.inv(E.T@(np.linalg.inv(C_qq))@E))

print('x_ML = '+str(round(x_ml[0][0],2)))
print('sigma_x_ML = '+str(round(sigma_x_ml[0][0],2)))
```

We found that the max likelihood solution for the mean MOC $\tilde x_{ML} \approx 17.02$ Sv and its standard error $\sigma_{\tilde x_{ML}} \approx 0.76$ Sv. I expect the $\tilde x_{ML}$ to be the most likely value of x given our observations of y. So in our case it is the most probable observed transport for the MOC.

### b. 
#### The number of effective degrees of freedom, $N_e$ , is defined in this problem such that the solution variance is equal to the sum of the MOC variance and the observational variance divided by $N_e$  

Solve for $N_e$ in this problem.

```python
'DEFINE VARIABLES'
q = (sigma_n**2) + (sigma_x**2) # variance of observational + random noise 
var_x_ml = np.linalg.inv(E.T@(np.linalg.inv(C_qq))@E) # solution variance 

'COMPUTE N_E EFFECTIVE DEGREES OF FREEDOM'
N_e = (q)/(np.sum(sigma_x_ml)**2) # variance of obs/noise over solution variance

print('N_e = ',str(round(N_e)))
```

### c. 
#### Take into account the autocovariance in the seasonal cycle. Assume that the statistics are stationary and the autocorrelation function is,

$\rho(\delta) = e^{(-[\delta/\tau_1]²)} cos(2π\delta / 12 [mon]),$

#### where $\tau_1 = 4 mon$. Solve for the mean, its standard error, and the number of degrees of freedom. The “toeplitz” function in MATLAB and the Julia package ToeplitzMatrices.jl are useful. (4 points)

From the notes, we know that the correlation between lagged elements can be described as $<x(t)'x(t+\delta)'> = \sigma_x^2 \rho(|\delta|)$ where $\delta$ is the lag if the time-series is stationary. If the variables are evenly spaced in time then our covariance matrix becomes $C_{xx} = \sigma_x^2 R_{\rho}$ where $R_{\rho}$ is the correlation matrix. The noise covariance matrix $C_{nn}$ is not affected.

$R_{\rho} = \begin{bmatrix}
\rho(0) & \rho(\delta) & \rho(2\delta) & ... & \rho(N\delta)\\
\rho(\delta) & \rho(0) & \rho(\delta) & ... & \rho((N-i)\delta)\\
\rho(2\delta) & \rho(\delta) & \rho(0) & ... & ...\\
... & ... & ... & \rho(0) & ...\\
\rho(N\delta) & \rho((N-j)\delta) & ... & \rho(\delta) & \rho(0)\\
\end{bmatrix}$

Therefore, we can use the autocorrelation function given $\rho(\delta)$ to redefine the covariance matrix and solve for the mean, its standard error, and degrees of freedom again.

```python
'DEFINE VARIABLES'
tau_1 = 4 # month
R_rho = np.empty((y.size, y.size)) # initialize empty correlation matrix size NxN (173x173)

# fill the correlation matrix with rho(delta) values
for delta in range(y.size):
    rho_delta = (np.exp(-((delta/tau_1)**2)))*(np.cos((np.pi*2*delta)/(12))) # autocorrelation function
    pos_diag = R_rho[delta:,:]
    neg_diag = R_rho[:,delta:]
    np.fill_diagonal(pos_diag,rho_delta)
    np.fill_diagonal(neg_diag,rho_delta)

'COMPUTE SOLUTION COVARIANCE AND NOISE COVARIANCE MATRICES'
C_xx = (sigma_x**2)*R_rho
C_nn = np.diag(np.repeat(sigma_n**2, y.size))

'RECOMPUTE THE OBSERVATIONAL + RANDOM NOISE MATRIX C_QQ'
C_qq = C_xx + C_nn

'REDO X_ML AND STANDARD ERROR CALCULATIONS FROM 1A'
x_ml = (np.linalg.inv(((E.T@(np.linalg.inv(C_qq))@E))))@((E.T@(np.linalg.inv(C_qq)))@y) # modified eqn 2.79
sigma_x_ml = np.sqrt(np.linalg.inv(E.T@(np.linalg.inv(C_qq))@E)) # modified eqn 2.82 

print('x_ML = '+str(round(x_ml[0][0],2)))
print('sigma_x_ML = '+str(round(sigma_x_ml[0][0],2)))
```

## Problem 2. Significance of correlation coefficient (10 points)

Two temperature timeseries have a correlation coefficient of $0.65$. Both timeseries are described by an autocorrelation function of

$ρ(\delta) = e^{(-[\delta/\tau]^2)}$,

where δ is the lag in days and the de-correlation timescale, $\tau$, is $20$ days. The expected variance of both timeseries is $1.0(°C)^2$.

### a) 
#### For a timeseries with record length, $T=100$ days and a resolution of $\Delta t=1$ day, compute the autocovariance matrix, $C_{xx}$. (2 points)

```python
'DEFINE THE VARIABLES'
T = 100
delta_t = 1 # resolution of the time-series
tau = 20
sigma = 1
corr_coeff = 0.65 # rho
var_x_expected = 1

'FIRST COMPUTE THE CORRELATION MATRIX R_RHO'
R_rho = np.empty((T, T)) # initialize empty correlation matrix size NxN (173x173)

# fill the correlation matrix with rho(delta) values
for delta in range(T):
    rho_delta = np.exp(-(delta/tau)**2) #autocovariance matrix
    pos_diag = R_rho[delta:,:]
    neg_diag = R_rho[:,delta:]
    np.fill_diagonal(pos_diag,rho_delta)
    np.fill_diagonal(neg_diag,rho_delta)

'COMPUTE THE AUTOCOVARIANCE MATRIX C_XX'
C_xx = (sigma_x**2)*R_rho

print('preview of C_xx = ')
print()
print(C_xx[0:4,0:4])
```

### b) 
#### Compute the Cholesky decomposition of $C_{xx}$. Use a dedicated function from your favorite programming language to do the decomposition. Add a small constant to all of the diagonal elements if the matrix is not positive definite. (1 point)

```python
np.fill_diagonal(C_xx, np.diag(C_xx)[0]+0.0001) # add small const. to diag elements of C_xx to make it pos. def.

'USE NP.LINALG.CHOLESKY TO GET CHOLESKY DECOMP OF C_XX'
L = np.linalg.cholesky(C_xx)

print('A preview of L = ')
print()
print(L[0:4,0:4])
```

$L$ is the Lower-triangular Cholesky factor of $a$ where $a$ is the Hermitian (symmetric if all elements are real), positive-definite input matrix.

### c) 
#### Create and plot two randomly generated timeseries with the autocovariance given above. (2 points)

### e) 
#### Is a correlation coefficient of $0.65$ different from zero at the $95%$ confidence level (i.e., $5%$ insignificance level)? Evaluate the null hypothesis that a coefficient of $0.65$ is due to chance. Do you accept or reject the null hypothesis? What is the $p$-value of the test. (4 points)

Our null hypothesis is that a coefficient of 0.65 is due to chance.

We will define our test statistic, $J$, to be related to the negative-log-likelihood function in (2.163), and find its expected value. The $J$ statistic is our goodness-of-fit statistic.

You reject the null hyptohesis when your value lies outside of the confidence interval.
The confidence interval is the interval in which 95% of your values in your distribution reside. 
An empirical CDF is comparable to confidence intervals. If you measure the probablilty that rho <= 0.65 and if that is >95% then you reject the null hypothesis. 


```python
'EMPIRICAL CDF FUNCTION'
def ecdf(x):
    xs = np.sort(x)
    ys = np.arange(1, len(xs)+1)/float(len(xs))
    return xs, ys

xs, ys = ecdf(corr_coeffs)

'PLOT THE CDF'
# create figure
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('CDF', fontsize=14, y=0.925, fontweight='bold')

# plot CDF
ax.plot(xs, ys);

# axes formatting
ax.set_xlabel('Correlation Coefficients', fontweight='bold')
ax.set_ylabel('Probability of rho being less than or equal to whatever value you give it',fontweight='bold');
ax.hlines(y=ys[np.where(xs<=0.65)[0][-1]],xmin=-1,xmax=1,color='k',linestyle='--', alpha=0.5);
ax.vlines(xs[np.where(xs<=0.65)[0][-1]],0,1,color='k',linestyle='--', alpha=0.5);
ax.hlines(y=ys[np.where(xs<=0.95)[0][-1]],xmin=-1,xmax=1,color='k',linestyle='--', alpha=0.8);
ax.vlines(xs[np.where(xs<=0.95)[0][-1]],0,1,color='k',linestyle='--', alpha=0.8);
ax.text(0, 0.92,'probability = 0.9036', fontsize = 7, color='grey');
ax.text(0, 1.015,'probability = 0.95', fontsize = 7, color='k');
```

We have found that the probability that $\rho >= 0.65$ is > the 95% confidence limit. This means we DO NOT reject the null hypothesis, therfore we cannot conclude that a coefficient of 0.65 is not due chance. In short, there is not enough evidence to support the claim that 0.65 is due to chance.

## Problem 3. Objective mapping (13 points)
Please refer back to the time series data that came with PSET #3: NewportPortlandSealevel.mat (or NewportPortlandSealevel.csv), and use the Newport, RI sea level data (column 3 of MATLAB variable).

Use Objective Mapping to estimate Newport sealevel for the first 25 days of data at 0.1 day intervals $(t= 1464:0.1:1489)$. This is a problem in one-dimensional objective mapping of a scalar variable. Perform the following steps.

### a. 
#### Make a covariance matrix that is a function of the lag, $\tau$, and has the following form:

$C_{xx} = 100 e^{[-( τ / 4)²]} \cdot cos(2πτ/8).$

The expected value for the sealevel is: $<𝐱> = 0.$


```python
'DEFINE VARIABLES'
tau_vec = np.arange(1464.5,1488.6,0.1)

'FIRST COMPUTE THE CORRELATION MATRIX C_xx'
C_xx = np.empty((tau_vec.size, tau_vec.size)) # initialize empty correlation matrix size NxN (173x173)

# fill the correlation matrix with rho(delta) values
for tau in range(tau_vec.size):
    diag = 100*np.exp(-1*((0.1*tau)/4)**2)*np.cos(2*np.pi*(0.1*tau)/8)
    pos_diag = C_xx[tau:,:]
    neg_diag = C_xx[:,tau:]
    np.fill_diagonal(pos_diag,diag)
    np.fill_diagonal(neg_diag,diag)

print('A preview of C_xx = ')
print()
print(C_xx[0:4,0:4]) # 241x241
```


### b. 
#### Form the noise covariance matrix: $C_{nn} = 10I$. (1 point)

```python
'DEFINE VARIABLES'
I = np.identity(25) # since we have 25 observations

'COMPUTE NOISE COVARIANCE MATRIX C_NN'
C_nn = 10*I

print('A preview of C_nn = ')
print()
print(C_nn[0:4,0:4]) #25x25

C_nn.shape
```

### c. 
#### Make the observational mapping matrix, 𝐄, that relates the gridded values, 𝐱, to each observation (3 points):

$𝐲 = 𝐄𝐱 + 𝐧.$

For the first two rows of $E$, for example, use equations of the form:

$y(t = 1464.5) = x( t = 1464.5) + n( t = 1464.5),$
$y(t = 1465.5) = x( t = 1465.5) + n( t = 1465.5).$

```python
# for gap filling, we want E such that it has 0s everywhere EXCEPT 1s where we have observations
# i have to really credit my collaborators here (Lucas & Oak), i wouldn't have come up with this nested for loop
# nearly as fast or made as efficient if they hadn't taught me how to do this part 


'DEFINE VARIABLES'
y_obs = n_slv[:25] # observations
t_obs = time_slv[:25]
E = np.empty((25, 241)) 

'CREATE E'
for i in range(25): # observations
    for j in range(241): # gap-filling
        if np.abs(t_obs[i]-tau_vec[j])<0.01: # when the diff when obs & estimate is < 0.1 then this is an observational data point!
            E[i,j]=1  # fill with a 1 for the observational data 
        else:
            E[i,j]=0 # fill with a 0 for the gap-filled data
            
print('A preview of E = ')
print()
print(E[0:12,0:4]) #25x241
```

### d. 
#### Make the common assumption that $C_{xn} = 0$, and therefore you can calculate (2 points):

$C_{xy} = C_{xx}E^{T}$

Now you can also calculate:


$C_{yy} = EC_{xx}E^T + C_{nn}$

```python
'CALCULATE COVARIANCE MATRICES C_XY AND C_YY'
C_xy = C_xx@E.T
C_yy = ((E@C_xx)@E.T) + C_nn

print('A preview of C_xy = ')
print()
print(C_xy[0:4,0:4]) #241x25

print()

print('A preview of C_yy = ')
print()
print(C_yy[0:4,0:4]) #25x25
```

### e. 
#### Calculate the objectively mapped estimate:

$\hat x = C_{xy}C_{yy}^{-1}y$

where the data has an expected value of zero $(<𝐲> = 0)$. Remove the sample mean of the data for consistency (2 points). Then add this value back into the solution.

```python
'DEFINE VARIABLES'
y_0 = y_obs - y_obs.mean() # remove the sample mean for consistency

'CALCULATE X_HAT'
x_hat_w_o_mean = (C_xy@np.linalg.inv(C_yy))@y_0
x_hat_w_mean = x_hat_w_o_mean + y_obs.mean()

print('A preview of x_hat with mean removed = ')
print()
print(x_hat_w_o_mean[0:4]) #241x1

print()

print('A preview of x_hat with mean = ')
print()
print(x_hat_w_mean[0:4]) #241x1
```

### f. 
#### Calculate the uncertainty of the estimate (2 points):

$P = C_{xx} - C_{xx}E^T(C_{yy}^{-1}EC_{xx})$

```python
'CALCULATE P UNCERTAINITY OF THE ESTIMATE'
P = C_xx - (C_xx@E.T)@((np.linalg.inv(C_yy))@E@C_xx)

'CALCULATE ERROR'
sigma_x = np.sqrt(np.diag(P))

print('A preview of P = ')
print()
print(P[0:4, 0:4]) # 251x251
```

### g. 
#### Plot the results, including the observations and final estimate, all with errorbars (4 points).

```python
'PLOT RESULTS'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Newport Sealevel Timeseries', fontsize=14, y=0.925, fontweight='bold')

# plot observations
ax.scatter(t_obs,y_obs, label='Observations', color='#fc8d59')
ax.errorbar(t_obs,y_obs,color='#fc8d59', yerr=np.sqrt(10), fmt="o",markersize=6) # plot observational data with standard error bars
# plot objective mapping estimate
ax.plot(tau_vec,x_hat_w_o_mean, label='Objective Mapping Estimate', color='navy')
ax.fill_between(tau_vec,(x_hat_w_o_mean.flatten() - sigma_x.flatten()), (x_hat_w_o_mean.flatten() + sigma_x.flatten()), color='navy', alpha=0.2, label= '1\u03C3 error')

# axes formatting
ax.set_xlabel('Time since Jan-1-1995 [year day]', fontweight='bold')
ax.set_ylabel('Adjusted sea level [cm]',fontweight='bold');
ax.legend();
ax.set_xlim([1464,1489])
ax.hlines(y=0,xmin=1464,xmax=1490,color='k',linestyle='--', alpha=0.5);
```
