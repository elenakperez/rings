### a) 
#### Compute the spatial EOFs of alongshore currents  using 1) the covariance approach and 2) the SVD approach. Show that the results are equivalent.

```python
v.shape
```

```python
'1) THE COVARIANCE APPROACH'

y = v # knowns are the along-shore current velocities, v
y = y - v.mean(axis = 1, keepdims = True) # we must remove the mean so that <y>=0, but do it for each column (axis=1)
C_yy = (y@y.T)/len(y[1]) # calculate the covariance matrix <y*y.T>
cov_eigval, cov_eigvec = np.linalg.eig(C_yy) # we want to find the eigenvectors of C_yy, which are the EOFs of U

print("Covariance approach eigenvectors =", cov_eigvec)
```

```python
'2) THE SVD APPROACH'

(U,lambdas,Vh)  = np.linalg.svd(y) # use numpy linalg to take SVD 
Lambda = np.diag(lambdas)
svd_eigvec = U # the U matrix are the 

print("SVD approach eigenvectors =", svd_eigvec)
print("")
print("Difference between covariance and SVD approach =",np.absolute(svd_eigvec) - np.absolute(cov_eigvec))
```

We can see that the differences between the two methods are essentially zero and likely come down to computational limits or roundoff error. Thus, the results are equivalent.

### e)
#### Compute wind stress from the winds, using $\rho_{air} = 1.3 \frac{kg}{m^3}$, $C_d = 1.4 \times 10^{-3}$, and: 
$\tau_x = \rho_{air} c_d (\sqrt{U^2 +V^2})U $

$\tau_y = \rho_{air} c_d (\sqrt{U^2 +V^2}) V$

#### Compute an EOF for the wind stress vector (using an approach where you treat the components of the vector as scalars). Compare the timeseries of EOF1 of the meridional velocity to EOF1 of the wind stress vector

```python
'SET UP WIND STRESS PROBLEM'
# knowns
rho_air = 1.3 
C_d = 1.4*(10**-3)

# subtract the mean from wind such that <U>=0, <V>=0
U = u_wind - u_wind.mean() 
V = v_wind - v_wind.mean()

# calculate wind stress using the formula
tau_x = rho_air*C_d*((np.sqrt(U**2 + V**2)))*U
tau_y = rho_air*C_d*((np.sqrt(U**2 + V**2)))*V

# create tau matrix
tau = np.squeeze(np.array([tau_x, tau_y]), axis = 1)

'USE SVD APPROACH TO COMPUTE EOF MODES'

# first, compute EOF modes for wind stress vector tau
(U_tau,lambdas_tau,Vh_tau)  = np.linalg.svd(tau) # use numpy linalg to take SVD 
Lambda_wind = np.diag(lambdas_tau)
b_tau = U_tau @ tau


'PLOT EOF 1 OF WIND STRESS AGAINST EOF 1 OF MERIDIONAL VELOCITY AT 5 M DEPTH'

# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Wind stress vector vs. meridional velocity @ 5 m depth', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(time[0],(b_tau[0] / np.sum(np.abs(b_tau))), label='Wind Stress EOF 1', color='#fc8d59')
ax.plot(time[0],y[0] / np.sum(np.abs(y)), label='Meridional Velocity EOF 1', color='navy')


# axes formatting
ax.set_xlabel('Time since Jan-1-1992 [year day]', fontweight='bold')
ax.set_ylabel('Beta (normalized)',fontweight='bold');
ax.legend();
ax.set_xlim([25,148])
ax.hlines(y=0,xmin=-1,xmax=200,color='k',linestyle='--', alpha=0.5);
```

## Problem 2. Consistency of observations.
Suppose we have two observations, yâ‚= 14Â°C Â± 1Â°C and yâ‚‚ = 12Â°C Â± 1Â°C, of one quantity, x. The errors are independent and well described by a normal distribution with the standard deviation given by the error bar.

### a) 
#### Solve and plot the likelihood function, the maximum likelihood solution, the standard error of the solution, and the 95% confidence limits.

```python
# observations
y_1 = 14
y_2 = 12
y_err = 1

# computed values
x_tilde = (((y_err**2)*y_2) + ((y_err**2)*y_1))/((y_err)**2 + (y_err)**2)
sigma_x = ((y_err)**2 * (y_err)**2)/(y_err**2 + y_err**2)

# confidence levels
ll_cl = x_tilde + 1.96*np.sqrt(sigma_x)
ul_cl = x_tilde - 1.96*np.sqrt(sigma_x)
```

$L(X|Y=y_1) = \frac{1}{\sqrt(2\pi)\sigma_n}e^{-\frac{(y_1-X)^2}{2\sigma_n^2}}$

$L(X|Y=y_2) = \frac{1}{\sqrt(2\pi)\sigma_n}e^{-\frac{(y_2-X)^2}{2\sigma_n^2}}$

Where $\sigma_n = \sigma_1 = \sigma_2$. The likelihood function is given by:

$p_{y|x}(Y|X)=L(X|Y=y_1)\cdot L(X|Y=y_2)$

Solve for the likelihood function:

$p_{y|x}(Y|X) = \frac{1}{\sqrt(2\pi)\sigma_n}e^{-\frac{(y_1-X)^2}{2\sigma_n^2}}\cdot \frac{1}{\sqrt(2\pi)\sigma_n}e^{-\frac{(y_2-X)^2}{2\sigma_n^2}}$

$ = \frac{1}{\sqrt(2\pi)\sigma_n}^2e^{-\frac{(y_1-X)^2}{2\sigma_n^2} -\frac{(y_2-X)^2}{2\sigma_n^2}}$

$p = \frac{1}{2\pi\sigma_n^2}e^{\frac{-(y_1-X)^2-(y_2-X)^2}{2\sigma_n^2}}$

Take the negative log of both sides to get rid of the negative exponential

$-log[P] = -log[\frac{1}{2\pi\sigma_n^2}e^{\frac{-(y_1-X)^2-(y_2-X)^2}{2\sigma_n^2}}]$

$-log[P] = log[2\pi\sigma_n^2] + \frac{(y_1-X)^2-(y_2-X)^2}{2\sigma_n^2}$

This is the log likelihood function. 

$[\frac{\partial(-log[P])}{\partial X}]_{x_{ML}} = \frac{\partial}{\partial X}(log[2\pi\sigma_n^2] + \frac{(y_1-X)^2-(y_2-X)^2}{2\sigma_n^2})$

Rewrite the expression with $\sigma_n = \sigma_1 = \sigma_2$.

$[\frac{\partial(-log[P])}{\partial X}]_{x_{ML}} = \frac{\partial}{\partial X}(log[2\pi\sigma_n^2] + \frac{\sigma_2^2(y_1-X)^2-\sigma_1^2(y_2-X)^2}{\sigma_1^2 + \sigma_2^2})$

To minimize, take the derivative w.r.t. X and set equal to 0.

$0 = \frac{\sigma_2^2 y_1 - \sigma_2^2 \textbf{x} + \sigma_1^2 y_2 - \sigma_1^2 \textbf{x}}{\sigma_1^2 + \sigma_2^2}$

Rearranging for $\textbf{x}$ and $\sigma_{\textbf{x}}$ we find:

$\textbf{x} = \frac{\sigma_2^2 y_1 + \sigma_1^2 y_2}{\sigma_1^2 + \sigma_2^2}$

$\sigma_{\textbf{x}} = \frac{\sigma_1^2 \sigma_2^2}{\sigma_1^2 + \sigma_2^2}$

Solving for the $\textbf{x}$ and $\sigma_{\textbf{x}}$ with our knowns we get:

$\textbf{x} = \frac{(1)(14) + (1)(12)}{(1)^2 + (1)^2} = 13 \deg C$

$\sigma_{\textbf{x}} = \frac{(1)^2 \cdot (1)^2}{(1)^2 + (1)^2} = 0.5 \deg C$

Therefore our $\textbf{x} = 13 \deg C \pm 0.5 \deg C$ where $13 \deg C$ is the maximum likelihood solution and $0.5 \deg C$ is its standard error. 

The confidence limits are given by $\textbf{x} \pm 1.96\sqrt{\sigma_{\textbf{x}}}$. The confidence level, $1 âˆ’ 2\alpha$, is $\approx 1.96$ for a typical choice $\alpha = 0.025$ such that the confidence level is $0.95$ or $95$%.

* lower confidence level: $\textbf{x} + 1.96\sqrt{\sigma_{\textbf{x}}} = 11.6 \deg C$

* upper confidence level: $\textbf{x} - 1.96\sqrt{\sigma_{\textbf{x}}} = 14.4 \deg C$

Meaning that if we were to take some samples, then we have a 95% confidence that our samples will be in this range.



```python
# create x for plotting
x = np.linspace(9, 17, 1000)

prod = (np.exp(-(y_1 - x)**2 / (2 * y_err**2)))*(np.exp(-(y_2 - x)**2 / (2 * y_err**2)))
norm = 1 / np.trapz(prod, x)
likelihood = norm * prod

'PLOT THE LIKELIHOOD FUNCTION, ML SOLN, THE STANDARD ERROR, AND 95% CONFIDENCE INTERVALS'

# create figure
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('PDF for consistency of observations problem', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(x, likelihood, color='#d73027', label='Likelihood Function') 
ax.plot([13-0.5, 13-0.5], [0, 0.445], c = 'orange', label='Standard error')
ax.plot([13+0.5, 13+0.5], [0, 0.445], c = 'orange')
ax.plot([ll_cl, ll_cl], [0, 0.075], c = 'blue', label='95% confidence level')
ax.plot([ul_cl, ul_cl], [0, 0.075], c = 'blue')
ax.plot([x_tilde, x_tilde], [0, 0.565], c = 'green',label='x_tilde')

# axes formatting
ax.set_xlabel('', fontweight='bold')
ax.set_ylabel('',fontweight='bold');
ax.legend();
# ax.set_xlim([11,15]);
# ax.hlines(y=0,xmin=-1,xmax=200,color='k',linestyle='--', alpha=0.1);
```

```python
# check the area under the curve is ~= 1
print("Area under the curve =",np.trapz(likelihood,x))
```

### b) 
#### Evaluate the null hypothesis that the two observations are consistent (i.e., that they measure the same physical quantity) at the 5% insignificance level using a two-sided chi-squared goodness-of-fit test. Calculate the p-value of the test using the following steps.

â€¢ choose $\alpha$, the level of significance, and find the region of acceptance defined by the upper and lower bounds, $\chi_U$ and $\chi_L$, where $P(x_L) = \alpha$ and $P(x_U) = 1-\alpha$ and $P$ is the expected cumulative distribution function of the test statistic

â€¢ vary $\alpha$ until the sampled test statistic, $\textbf{J}$ has a value at the edge of the region of acceptance

â€¢ the probability that the hypothesis is wrongly rejected is the p-value and is equal to $2\alpha$

Our null hypothesis is that the two observations $y_1, y_2$ are consistent. First, we will state that our test statistic J:

$J = \frac{(y_1 -x_{true})^2}{\sigma_1^2} + \frac{(y_2 -x_{true})^2}{\sigma_2^2}$

Second, we will compute the value of J_tilde.

```python
'COMPUTE <J_tilde>'

M = 2 # number of obs
N = 1 # number of constraints
degoff = M-N # degrees of fredom = number of obs - number of constraints

# obs
y_1 = 14
y_2 = 12
y_err = 1

# computed values
x_tilde = (((y_err**2)*y_2) + ((y_err**2)*y_1))/((y_err)**2 + (y_err)**2)
sigma_x = ((y_err)**2 * (y_err)**2)/(y_err**2 + y_err**2)

# std
sigma_1 = 1
sigma_2 = 1

# test statistic
J_tilde = ((y_1 - x_tilde)/sigma_1)**2 + ((y_2 - x_tilde)/sigma_2)**2
print("J_tilde =", J_tilde)
```

Our expected value of $<J_{tilde}> = \nu = M-N = 1$. Next, we will decide the power of the test. For $\alpha$ we choose the 5% significance level. From the notes, we know the 95% confidence interval for the $\chi_1^2$
variable is $\chi_L = 0.001$ and $\chi_U = 5.0$.

Since our test statistics J_tilde = 2 and is in this interval (2 is between 0.001 and 5). Therefore, our null hypothesis is NOT rejected. The two data points are accepted as consistent at the 5% insignificance (95% confidence) level.

```python
# create variance function to calculate variances of the time-series 
def variance_func(data):
    n = len(data)
    mean = sum(data) / n     # Mean of the data
    deviations = [(x - mean) ** 2 for x in data]     # Square deviations
    variance = sum(deviations) / n # Variance
    return variance
```

```python
'CALCULATE THE EXPECTED VALUES OF THE NEWPORT AND PORTLAND SEA LEVEL FLUCUATIONS'

JPF_mean = ((n_slv - n_slv.mean())*(p_slv - p_slv.mean())).mean()
                
print("The expected value of Newport sea level flucuations is = ",round(variance_func(n_slv)[0],1))
print("The variance of Portland sea level flucuations is = ",round(variance_func(p_slv)[0],1))
print("The variance of the Newport and Portland sea level flucuations = ", round(JPF_mean, 1))
```

The expected value of the joint PDF for Newport and Portland is on the order of variance for each individual city's sea level flucuations.
