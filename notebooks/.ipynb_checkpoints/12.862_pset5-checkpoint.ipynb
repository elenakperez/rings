## 1. Dispersion relationship…or how shallow is shallow?
### b. Plot the non-dimensional dispersion vs.  kh

```python
'DEFINE VARIABLES'
g = 9.8 # m/s^2 , gravitational acceleration constant
kh = np.arange(0,1,0.001) # k*h [unitless]
h_1 = 1 # m , shallow water
h_10 = 10 # m , deeper water

h = 1

'NON-DIMENSIONAL DISPERSION RELATIONSHIP --> (((omega^2)*h)/g) = kh*(np.tanh(kh))'
nondim_eqn = kh*(np.tanh(kh)) 
# shallow_water_approx_eqn = (((kh)**2)/(h**2))*h*g # np.tanh(kh) ~= 1

shallow_water_approx_eqn = (kh)**2

# deep_water_approx_eqn = kh

'FIND WHERE SHALLOW WATER APPROX. IS GREATER THAN 20% ERROR FROM NON-DIM DISP. RELATIONSHIP'
error20 = kh[np.where(((np.abs(nondim_eqn - shallow_water_approx_eqn)/nondim_eqn) >= 0.2))[0][0]]
```

```python
'PLOT THE RELATIONSHIP'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Non-dimensional Dispersion Relationship vs.  kh', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(kh,nondim_eqn,label='Non-dimensional Dispersion Rel.')
ax.plot(kh,shallow_water_approx_eqn,label='Shallow Water Approx.')

# axes formatting
ax.set_xlabel('kh', fontweight='bold')
ax.set_ylabel('$\omega^2/hg = kh\cdot tanh(kh)$',fontweight='bold')
ax.legend();
ax.set_ylim([-0.1,1])
ax.set_xlim([0,1])
ax.vlines(error20,-1,3,color='k',linestyle='--', alpha=0.1);
# ax.hlines(0,-1,3,color='k',linestyle='--', alpha=0.1);
ax.text(error20-0.275, 0.8, 'kh = '+str((round(error20,2)))+' is where shallow water approx. = 20% error', fontsize = 7, color='grey');
```

### c. For h = 10 m, plot f versus k for the full and shallow water dispersion relationship.

```python
'DEFINE NEW VARIABLES'
g = 9.8 # m/s^2 , gravitational acceleration constant
kh = np.arange(0,1,0.001) # k [1/m], wavenumber
h = 10 # [m], water depth given in problem statement
k = kh/h

'EQUATIONS'
full_f = (np.sqrt(g*k*np.tanh(kh)))/(2*np.pi)
sw_f = (np.sqrt(g*(k**2)*h))/(2*np.pi)

'FIND WHERE SHALLOW WATER APPROX. IS GREATER THAN 10% ERROR FROM NON-DIM DISP. RELATIONSHIP'
error10 = k[np.where((((np.abs(full_f - sw_f))/full_f) >= 0.1))[0][0]]
```

```python
'PLOT THE RELATIONSHIP'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('f vs. k for non-dimensional full & shallow water disp. relations', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(k,full_f,label='Full Dispersion Relationship')
ax.plot(k,sw_f,label='Shallow Water Dispersion Relationship')

# axes formatting
ax.set_xlabel('k', fontweight='bold')
ax.set_ylabel('$f = \omega^2/2\pi$',fontweight='bold')
ax.legend();
ax.set_ylim([0,0.16])
ax.set_xlim([0,0.1])
ax.vlines(error10,-1,18,color='k',linestyle='--', alpha=0.1);
# ax.hlines(0,-1,3,color='k',linestyle='--', alpha=0.1);
ax.text(error10-0.0275, 0.14, 'k = '+str((round(error10,2)))+' is where shallow water approx. = 10% error', fontsize = 7, color='grey');
```

## 4. Stokes drift
### c. For shallow-water waves in water depth h = 10 m and with wave height H = 1 m, what is u_s  at z=0? Plot the vertical profile. What is the depth-averaged Eulerian flow?

```python
'CREATE THE VARIABLES'
h = 10 # m, bottom depth
z = np.arange(0,h,0.001)
H = 1 # m, wave height
g = 9.8 # m/s, gravitational constant
c = np.sqrt(g*h) # shallow water group/phase speed
a = 0.5*H
f = 0.01 # 1/s 
k = (f*2*np.pi)/c
u_s = ((c*(a**2))/(2*h**2))*(np.cosh(2*k*z)) # for z = 0 in shallow water
```

```python
'PLOT THE RELATIONSHIP'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Stokes drift for shallow water waves h = 10 m & H = 1 m', fontsize=14, y=0.925, fontweight='bold')

# plot
ax.plot(u_s,z,label='Vertical Profile')

# axes formatting
ax.set_xlabel('u_s', fontweight='bold');
ax.set_ylabel('$z$',fontweight='bold');
```
