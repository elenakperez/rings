## EOF Analysis

```python
P3_prs.shape

P1_prs = P1_prs[~np.isnan(P1_prs)]
P2_prs = P2_prs[~np.isnan(P2_prs)]
P3_prs = P3_prs[~np.isnan(P3_prs)]

P1_prs = (np.reshape(P1_prs, (len(P1_prs),1)))[:min(len(P1_prs), len(P2_prs), len(P3_prs))]
P2_prs = (np.reshape(P2_prs, (len(P2_prs),1)))[:min(len(P1_prs), len(P2_prs), len(P3_prs))]
P3_prs = (np.reshape(P3_prs, (len(P3_prs),1)))[:min(len(P1_prs), len(P2_prs), len(P3_prs))]


P12_prs = (np.concatenate((P1_prs, P2_prs),1))
P123_prs = (np.concatenate((P12_prs, P3_prs),1)).T

P123_prs.shape
```

```python
P123_prs[~np.isnan(P123_prs)].shape
```

```python
y = P123_prs

y = P123_prs - P123_prs.mean(axis = 1, keepdims = True)

'THE SVD APPROACH'
(U,lambdas,Vh)  = np.linalg.svd(y) # use numpy linalg to take SVD 
Lambda = np.diag(lambdas)
svd_eigvec = U # the U matrix are the 

'COMPUTE PERCENT VARIANCE EXPLAINED BY EOF 1'
percent_variance_1 = (Lambda[0]**2) / (np.sum(Lambda[:]**2))
print("The percent variance explained by EOF 1 is", round(percent_variance_1[0]*100,1),"%")

'COMPUTE EXPANSION COEFFICIENTS'
b = U.T@y # to find the expansion coeff use the formula from the notes b = U.T@y
b1 = b

'CALCULATE PERCENT VARIANCE EXPLAINED BY THE EOF MODES'
percent_variance = []
for i in range(len(Lambda)):
    percent_variance.append((Lambda[i]**2) / (np.sum(Lambda[:]**2)))
    
print("The percent explained by the EOFs in descending order is", np.diag(percent_variance)*100)



'PLOT EXPANSION COEFFICIENT TIMESERIES'
# create figure 
fig,ax = plt.subplots(figsize=(10,7))
fig.set_dpi(100)
fig.suptitle('Expansion coefficient timeseries for bottom pressure', fontsize=14, y=0.925, fontweight='bold')

# plot
# ax2=ax.twinx()
ax.plot(P1_taudd[:3572], b[0],label='EOF 1', color='#d73027') 
ax.plot(P1_taudd[:3572],b[1],label='EOF 2', color='#fc8d59')
ax.plot(P1_taudd[:3572], b[2],label='EOF 3', color='#fee090') 
# ax.plot(time[0],b[3],label='EOF 4', color='#91bfdb') 
# ax.plot(time[0],b[4],label='EOF 5', color='#4575b4') 

# axes formatting
ax.set_xlabel('Time since Jan-1-2022 [yearday]', fontweight='bold')
ax.set_ylabel('Beta',fontweight='bold');
ax.hlines(y=0,xmin=85,xmax=250,color='k',linestyle='--', alpha=0.1);
ax.set_xlim([89,240])
ax.set_ylim([-1,1])


# legend formatting
colors = ['#d73027', '#fc8d59', '#fee090']
lines = [Line2D([0], [0], color=c, linewidth=3) for c in colors]
labels = ['EOF 1', 'EOF 2', 'EOF 3']
plt.legend(lines, labels, loc='upper right');
```

```python
def band_avg_autospec_dens_plot(fm_avg, spec_dens_avg, M, title, fig_quality, lineColor, confidColor, ifColor):
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    This function produces a plot of the autospectral density of a time sereis spec_dens_avg plotted against
    the frequency averaged fm_avg with a 95% confidence interval.
    
    Input:
        spec_dens_avg (array)       : band-averaged autospectral density of the time series x
        fm_avg (array)              : band-averaged frequency of the time series x
        title (string)              : title of the figure 
        fig_quality (int)           : quality of the figure, e.g. 100 dpi
        lineColor (string)          : color of the line plot
        confidColor (string)        : color of the confidence interval
        ifColor (bool)              : if True use custom colors, else False use default color scheme

    Output: 
    * returns plot of the band-averaged autospectral density of a time-series plotted against the frequencies
    
     
    """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    # create figure
    fig,ax = plt.subplots(figsize=(10,7))
    fig.set_dpi(fig_quality)
    fig.suptitle(title, fontsize=14, y=0.925, fontweight='bold')

    # plot
    if ifColor:
        ax.loglog(fm_avg, spec_dens_avg, color=lineColor, alpha=0.8);
        confidence_interval(0.05,M*2,confidColor)
    else:
        ax.loglog(fm_avg, spec_dens_avg, color='#2c7fb8', alpha=0.8);
        confidence_interval(0.05,M*2,'#253494')

    # axes formatting
    ax.set_xlabel('Frequency [1/hr]',fontweight='bold');
    ax.set_ylabel('Autospectral Density [(m/s)^/(1/s)]',fontweight='bold');
    
    return fig, ax
```
