# 1. Mean of an autocorrelated timeseries (14 pts)

Download the text file containing the Atlantic meridional overturning circulation (MOC) at 26.5°N, in monthly averages starting with April, 2004 (moc24N_monavg.txt). The standard error in the inference of monthly mean MOC is 0.1 Sv with an expected normal distribution.

```python
df_q1 = pd.read_csv('moc24N_monavg.txt',header=None) #import the file into a pandas dataframe
df_q1.columns =['MOC_avg'] #name the (one and only) column of the data
mean = df_q1.mean() #take the mean of the dataframe
df_q1_norm = df_q1-df_q1.mean(axis=0) #create a dataframe where the mean is subtracted out

#plot the timeseries for a better understanding of the data
plt.figure(figsize=(15,3))
plt.plot(df_q1['MOC_avg'], color = 'deepskyblue')
plt.title('Atlantic Meridional Overturning Circulation (MOC)')
plt.xlabel('Months from April 2004')
plt.ylabel('Sverdrup Transport [Sv]')
```

<b>1.a) Diagnose the maximum likelihood solution for the mean MOC and its standard error under the assumption that each monthly average is independent and the variance of the MOC is 100 Sv². You may have to derive a solution formula that is unique from any found in the class notes. (5 pts)</b>

given:
- normal distribuiton
- std err = 0.1 Sv = $\sigma_{n}$ (the square root of the solution covariance of the noise)
- variance of the $ MOC = 100 Sv^2  = \sigma_{x}^{2}$ (assuming independent average (monthly) obs)

want to find: 
* maximum likelihood solution for the mean MOC $\tilde{x}_{ML}$ and it's standard error $\sigma_{\tilde{x}_{ML}}$

work:
* $\tilde{x}_{ML}=(E^{T}C_{qq}^{-1}E)^{-1}E^{T}C_{qq}^{-1}y$ (eqn 2.79 with $C_{qq}$ substituted in for $C_{nn}$)
* recall that $C_{qq}=C_{xx}+C_{nn}$, and $C_{qq}$ is a diagonal matrix with the elements of the diagonal equal to $\sigma_{x}^{2}+\sigma_{n}^{2}$
* let $E$ = and Nx1 row-vector of ones (because we are letting the background state be constant (which Lilli suggested I do))
* let $C_{qq}$ be the diagonal matrix described above, plugging in our $\sigma_{n}$ and $\sigma_{x}$ from the given
* let $y$ be a 1xN column vector of the MOC averages in the dataframe
* also, recall that $\sigma_{\tilde{x}_{ML}}^{2}=(E^{T}C_{qq}^{-1}E)^{-1}$   (inverse of eqn 2.94)  which is the standard error
* alternatively, this is eqn 2.203 for the uncertainty of the solution $P=(E^{T}C_{qq}^{-1}E)^{-1}$

```python
#%% Problem 1a
E_1 = np.array([np.ones(len(df_q1))]).T #make this a column vector by transposing
Cqq_1 = np.zeros((len(df_q1), len(df_q1)), int) #create an empty matrix that we will fill in the next few lines
sigma_n_1 = 0.1 #from given
sigma_x_1 = 10  #from given
np.fill_diagonal(Cqq_1, ((sigma_n_1**2)+(sigma_x_1**2))) #filling our Cqq_1 matrix
y_1 = np.array([df_q1['MOC_avg']]).T #make this a column vector by transposing
# eqn 2.79 below
Xml_1a = np.matmul(np.matmul(np.matmul((np.linalg.pinv(np.matmul(np.matmul(E_1.T,np.linalg.pinv(Cqq_1)),E_1))),E_1.T),np.linalg.pinv(Cqq_1)),y_1)
# inverse of eqn 2.94 below (or eqn 2.203)
sigma_Xml_1a = np.sqrt(np.linalg.pinv(np.matmul(np.matmul(E_1.T,np.linalg.pinv(Cqq_1)),E_1)))

print(str(Xml_1a) + " +/- " + str(sigma_Xml_1a))
```

<b>1.b) The number of effective degrees of freedom, $N_e$, is defined in this problem such that the solution variance is equal to the sum of the MOC variance and the observational variance divided by $N_e$ in $\sigma_{\tilde{x}_{ML}}^{2} = (\sigma_{x}^{2}+\sigma_{n}^{2})/N_{e}$  .   Solve for $N_e$ in this problem (2 pts).</b>

```python
#from given equation in question:
Ne_1a = ((sigma_x_1**2)+(sigma_n_1**2))/(sigma_Xml_1a**2)
Ne_1a = math.floor(Ne_1a) #degrees of freedom must be a whole number, so we round down to the nearest whole nuber
print(str(Ne_1a))
```

<b>1.c) Take into account the autocovariance in the seasonal cycle. Assume that the statistics are stationary and the autocorrelation function is, $\rho(\delta) = exp(-[\frac{\delta}{\tau_1}]^{2})cos(\frac{2\pi\delta}{12 [mon]})$ , where $\tau_1=4 mon$. Solve for the mean (i), its standard error (ii), and the number of degrees of freedom (iii). (4pts)</b>

we will need to create a matrix for $R_\rho(\delta)$ which will be a square NxN matrix with $i$ as the index number of rows and $j$ as the index number of columns. 
$$
\begin{bmatrix} 
\rho(0) & \rho(\delta) & \rho(2\delta) & ... & \rho(N\delta) \\
\rho(\delta) & \rho(0) & \rho(\delta) & ... & \rho((N-i)\delta) \\
... & ... & ... & ... & ... \\
\rho(N\delta) & \rho((N-j)\delta) & ... & \rho(\delta) & \rho(0) \\
\end{bmatrix}
\quad
$$

The $R_\rho(\delta)$ matrix impacts the $C_{xx}$ matrix (below)
$$
\begin{bmatrix} 
\sigma_{x}^2 & 0 & ... &0 \\
0 & \sigma_{x}^2 & ... &0 \\
0 & 0 & ... &... \\
... & ... & ... &\sigma_{x}^2 \\
\end{bmatrix}
$$

which turns the diagonals into $\sigma_{x}^2R_\rho(\delta)$ (below)
$$
\begin{bmatrix} 
\sigma_{x}^2R_\rho(\delta) & 0 & ... &0 \\
0 & \sigma_{x}^2R_\rho(\delta) & ... &0 \\
0 & 0 & ... &... \\
... & ... & ... &\sigma_{x}^2R_\rho(\delta) \\
\end{bmatrix}
$$

The $C_{nn}$ matrix is unchanged:
$$
\begin{bmatrix} 
\sigma_{n}^2 & 0 & ... &0 \\
0 & \sigma_{n}^2 & ... &0 \\
0 & 0 & ... &... \\
... & ... & ... &\sigma_{n}^2 \\
\end{bmatrix}
$$

```python
tau_1 = 4 #from given
R_rho_1c = np.zeros((len(E_1), len(E_1)), float) #create a blank matrix to fill later

for delta in range(R_rho_1c.shape[0]): #this for loop will build the matrix R_rho(delta)
    diag_val_1 = np.exp(-1*(delta/tau_1)**2)*np.cos(2*np.pi*delta/12) #this is from the given equation for rho(delta)
    pos_diag_1 = R_rho_1c[delta:,:]
    neg_diag_1 = R_rho_1c[:,delta:]
    np.fill_diagonal(pos_diag_1, diag_val_1)
    np.fill_diagonal(neg_diag_1, diag_val_1)
Cxx_1c = (sigma_x_1**2)*R_rho_1c #from matrix above
Cnn_1c = np.zeros((len(df_q1), len(df_q1)), float) #blank matrix to fill below
np.fill_diagonal(Cnn_1c, (sigma_n_1**2)) #filling the Cnn matrix
Cqq_1c = Cxx_1c+Cnn_1c #creating the new Cqq matrix
# (i) # updated eqn 2.79 below
Xml_1c = np.matmul(np.matmul(np.matmul((np.linalg.pinv(np.matmul(np.matmul(E_1.T,np.linalg.pinv(Cqq_1c)),E_1))),E_1.T),np.linalg.pinv(Cqq_1c)),y_1)
# (ii) # updated inverse of eqn 2.94 below (or eqn 2.203)
sigma_Xml_1c = np.sqrt(np.linalg.pinv(np.matmul(np.matmul(E_1.T,np.linalg.pinv(Cqq_1c)),E_1)))
# (iii) # updated Ne equation
Ne_1c = ((sigma_x_1**2)+(sigma_n_1**2))/(sigma_Xml_1c**2)
Ne_1c = math.floor(Ne_1c)
print(str(Xml_1c)+" +/- "+str(sigma_Xml_1c))
print(str(Ne_1c))
```

Take into account model results that there is autocovariance in the MOC at decadal timescales. # 2. Significance of correlation coefficient (10 pts)

Two temperature timeseries have a correlation coefficient of 0.65. Both timeseries are described by an autocorrelation function of $\rho_\delta = exp(-[\frac{\delta}{\tau}]^2)$  where $\delta$ is the lag in days and the de-correlation timescale, $\tau$, is 20 days. The expected variance of both timeseries is $1.0(°C)^2$.

<b>2.a) For a timeseries with record length, $T=100 [day]$ and a resolution of $\Delta t=1 [day]$, compute the autocovariance matrix, $C_{xx}$. (2 points) </b>

We recall from problem 1 that an autocovariance $C_{xx}$ matrix is
$$
\begin{bmatrix} 
\sigma_{x}^2R_\rho(\delta) & 0 & ... &0 \\
0 & \sigma_{x}^2R_\rho(\delta) & ... &0 \\
0 & 0 & ... &... \\
... & ... & ... &\sigma_{x}^2R_\rho(\delta) \\
\end{bmatrix}
$$

so we do the same thing we did in 1c to find $C_{xx}$

```python
tau_2a = 20 #given
R_rho_2a = np.zeros((100,100), float) #create empty matrix to fill in the following for-loop
#for loop for creating/filling the R_rho(delta) matrix
for i in range(R_rho_2a.shape[0]):
    diag_val_1 = np.exp(-1*(i/tau_2a)**2)
    pos_diag_1 = R_rho_2a[i:,:]
    neg_diag_1 = R_rho_2a[:,i:]
    np.fill_diagonal(pos_diag_1, diag_val_1)
    np.fill_diagonal(neg_diag_1, diag_val_1)
sigma_x_2 = math.sqrt(1.0) #from given
Cxx_2a = (sigma_x_2**2)*R_rho_2a #create the Cxx matrix for autocorrelation

print(Cxx_2a)
```

<b>2.b) Compute the Cholesky decomposition of $C_{xx}$. Use a dedicated function from your favorite programming language to do the decomposition. Add a small constant to all of the diagonal elements if the matrix is not positive definite. (1 point)</b>

The Cholesky decomposition is: $C_{xx} = C_{xx}^{\frac{1}{2}T}C_{xx}^{\frac{1}{2}}$   and we will use the $C_{xx}$ we calculated in 2a for the RHS of this equation (the LHS is the Cholesky decomposed $C_{xx}$)
* we need a small constant added to the $C_{xx}$ matrix from 2a because we can't do the decomposition unless the $C_{xx}$ matrix is positive definite, which can be achieved by adding a VERY small constant to all the diagonal elements of the matrix. 
* I started out letting the small constant be $10^{-16}$, which was too small, and then I increased by powers of 10 until I reached a number that ran without giving the "Matrix is not positive definifte" error message anymore... this value was $10^{-14}$

```python
small_constant = 1*10**-14 #see explanation above for choice of this value
small_constant_matrix = small_constant*np.identity(len(R_rho_2a)) #we only need to change the diagonal elements so
# we make a new matrix of the same size as the Cxx so that we can add them together to add in the small constant
Cxx_2ab = ((sigma_x_2**2)*R_rho_2a) + small_constant_matrix #call this 2ab since it's inbetween step to get a new
# matrix from Cxx_2a that has the small constant added to it
Cxx_2b = np.linalg.cholesky(Cxx_2ab) # cholesky decomposition of the new Cxx matrix with the small constant

print(Cxx_2b)
```

<b>3.a) Make a covariance matrix that is a function of the lag, $\tau$, and has the following form:
$C_{xx}=100exp[-(\frac{\tau}{4})^2]cos(\frac{2\pi\tau}{8})$  . Let the expected value for the sealevel be $<x> = 0$.
</b>

```python
tau_3 = np.arange(1464,1489.1,0.1) #from given
Cxx_3 = np.zeros((len(tau_3), len(tau_3)), float)
for tau in range(tau_3.size):
    diag_val_3 = 100*np.exp(-1*((0.1*tau)/4)**2)*np.cos(2*np.pi*(0.1*tau)/8) #since tau is an integer, we need to 
    # multiply it by 0.1 to get the proper interval range representative of our tau_3 variable
    pos_diag_3 = Cxx_3[tau:,:]
    neg_diag_3 = Cxx_3[:,tau:]
    np.fill_diagonal(pos_diag_3, diag_val_3)
    np.fill_diagonal(neg_diag_3, diag_val_3)
print(Cxx_3)
```

<b>3.b) Form the noise covariance matrix: $C_{nn} = 10I$ (1 point)</b>

```python
Cnn_3 = 10*np.identity(25)

print(Cnn_3)
```

Lucas and Christian explained to me that E should be a 25 x 251 matrix (# of single day obs x # of estimated obs), and all the values should be zero except for 25 places where the value we are estimating is the value we have already observed in the once-a-day observations. <b>3.d) Make the common assumption that $C_{xn}=0$, and therefore you can calculate (2 points):

<b>3.e) Calculate the objectively mapped estimate:
